#### Neden Boyut İndirgemesi Yapıyoruz ? 

*Gerçek hayattaki veri setlerinde karşılaşacağımız yaygın 2 problem var: **Çoklu Doğrusal Bağlantı Problemi** ve **Çok Boyutluluk Problemi**. Bağımsız değişkenler arasında yüksek bağımlılık söz konusu olabilir yani yüksek korelasyon olabilir. Yani bir değişken başka bir değişkenin sebebi veya sonucu olabilir. Bu da EKK tahmincileri gibi model geliştiricilerin elde ettikleri parametrelerin yüksek yanlılığa sebebiyet vermesiyle sonuçlanır. Bir de bazı veri setlerinde değişken sayısının gözlem birimi sayısından fazla olması sorunu var.*

*PCR bu iki probleme de çözüm sunuyor. Örneğin 100 değişkenli bir veri setini, o veri setindeki bilgiyi büyük ölçüde temsil edebilecek 2-3 değişkene yani bileşene indirgeme işlemi uyguluyor. Burada indirgeme sonucunda çok az da olsa bilgi kaybı söz konusu ama göz ardı edilebilir. Zaten bileşen sayısında azalma olduğu için çok boyutluluk problemini de ortadan kaldırıyoruz. İndirgeme yaptığımız bileşenler arasında korelasyon ilişkisi olmayacak. Çoklu doğrusal problemini de bu şekilde ortadan kaldırmış oluyoruz.*
