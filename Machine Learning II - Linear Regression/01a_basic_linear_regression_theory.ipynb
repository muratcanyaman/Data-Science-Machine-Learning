## Basit Doğrusal Regresyon (Teori)

*Temel amaç: Bağımlı ve bağımsız değişkenler arasındaki ilişkiyi ifade eden doğrusal fonksiyonu bulmaktır.*

*Not: Basit Doğrusal Regresyon Makine Öğrenmesi Algoritmalarının çıkış noktasıdır. Aynı zamanda en temel, en kompleks olmayan algoritma olduğu için burada detaylı bir şekilde teorisini anlatmaya calışacağız. Yani amaç sağlam bir temel atmak. *Lineer regresyonda kullandığımız gösterim iki farklı seviyede ele alınıyor. Anakitle teorik gösterimi, tüm popülasyona ait gerçek parametreleri ifade ediyor: β₀ sabit terim, β₁ eğim katsayısı ve ε de hata terimi. Ancak bu değerlere pratikte hiçbir zaman doğrudan ulaşamıyoruz.*

*Bu yüzden elimizdeki örneklem verisini kullanarak bu parametreleri tahmin etmeye çalışıyoruz. Örneklem gösterimindeki b₀ ve b₁, işte bu bilinmeyen β₀ ve β₁ değerlerinin tahmincileridir.*

*Peki tahmin modeli neden ayrı gösterilmiş? Çünkü biz modeli kullanırken yalnızca b₀ ve b₁ ile bir tahmin üretiyoruz; hata terimini önceden bilemiyoruz. Hata terimi eᵢ, tahmin yaptıktan sonra ortaya çıkan bir gerçek-tahmin farkıdır.*

*Yukarıdaki görselleri biraz açalım. Hata terimi eᵢ = yᵢ - ŷᵢ, yani gerçek değer ile tahmin edilen değer arasındaki fark. Modelin ne kadar yanıldığını bu formülle ölçüyoruz.*

*SSE (Sum of Squared Errors) ise tüm bu hataların karelerinin toplamı. Hataları karelerine alarak hem negatif hem pozitif sapmaların birbirini götürmesini önlüyoruz. SSE formülüne tahmin modelini ŷᵢ = b₀ + b₁xᵢ yerleştirdiğimizde ifade tamamen b₀ ve b₁ cinsinden yazılmış oluyor.*

*İşte tam bu noktada modelin amacı netleşiyor: SSE'yi minimize edecek en optimum b₀ ve b₁ değerlerini bulmak. Bunun için SSE formülünün b₀ ve b₁'e göre kısmi türevleri alınıp sıfıra eşitlendiğinde, bu iki katsayıya doğrudan ulaşabiliyoruz.*

*Grafikteki mavi noktalar, veri setindeki gerçek bağımlı değişken değerlerini yani gerçek y değerlerini temsil ediyor. Kırmızı düz çizgi ise bizim tahmin fonksiyonumuz; modelin tahmin ettiği değerleri gösteriyor.*

*Şimdi grafikte bir xᵢ değeri düşünelim. Bu xᵢ değerini b₁ eğim katsayısıyla çarpıp b₀ sabit terimini eklediğimizde tahmin edilen ŷᵢ değerine ulaşıyoruz. Bir de hata payı olan eᵢ'yi dahil ettiğimizde gerçek yᵢ değerine ulaşmış oluyoruz. Yani formül şu şekilde özetleniyor: yᵢ = b₀ + b₁xᵢ + eᵢ*

*Peki buradaki eᵢ nedir? Modelin her bir xᵢ için yaptığı tahminin, gerçek değerden ne kadar saptığını gösteriyor. Yani grafikte düz çizgi üzerindeki tahmin ile mavi nokta arasındaki dikey fark, o noktanın hata payını veriyor.*

*Model dediğimiz şey aslında şu: Gerçek değerlere en yakın tahminleri yapabilmek için en optimum b₀ ve b₁ katsayılarını bulmak. Bu katsayıları ne kadar iyi ayarlarsak, hata payları da o kadar küçülüyor ve modelimiz o kadar başarılı oluyor.*
