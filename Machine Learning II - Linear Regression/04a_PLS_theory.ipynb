## Kısmi En Küçük Kareler Regresyonu (PLS - Teori)

*Değişkenlerin daha az sayıda ve aralarında çoklu doğrusal bağlantı problemi olmayan bileşenlere indirgenip regresyon modeli kurulması fikrine dayanır. Prensipte PCR'a benzer.*

*yukarıdaki diyagramda gösterildiği gibi PLS'de PCR'de olduğu gibi X adet bağımsız değişken ile oluşturulan Y adet bileşen ile bağımlı değişkenlerin tahmini işlemi gerçekleştirilir.*

*Bu algoritmayı kullanmamızdaki temel sebep yine PCR'da olduğu gibi çok doğrusallılığı ve çok boyutluluğu ortadan kaldırmak.*

- *Çok boyutluluk laneti p > n*<br>
- *Çoklu doğrusal bağlantı problemi*<br>
- *PLS de PCR gibi bağımsız değişkenlerin doğrusal kombinasyonlarını bulur. Bu doğrusal kombinasyonlar bileşen ya da latent değişken olarak adlandırılır.*<br>
- *PLS NIPALS'in özel bir halidir, iteratif olarak bağımlı değişken ile yüksek korelasyona sahip değişkenler arasındaki gizil(latent) ilişkiyi bulmaya çalışır.*

### PLS vs PCR

- *PCR'da doğrusal kombinasyonlar yani bileşenler **bağımsız değişken uzayındaki değişkenliği**  maksimum şekilde özetleyecek şekilde oluşturulur. Bu durum bağımlı değişkeni açıklama yeteneği olmamasına sebep olmakta. Yani indirgeme bağımsız değişkenler arasındaki o korelasyon bağlantısını kırıyor fakat bu işlem bağımlı değişkeni açıklama yeteneği olmayan bir işlem.*

- *PLS'de ise **bileşenler bağımlı değişken ile olan kovaryansı(iki değişkenin birlikte ne yönde ne şidette değişmesi)** maksimum şekilde özetleyecek şekilde bir işlem yapılır.*

- *Yani değişkenler atılmak istenmiyorsa ve açıklanabilirlik aranıyorsa: PLS*<br>

- *PLS, gözetimli boyut indirgemesi, PCR ise gözetimsiz boyut indirgemesi olarak görülebilir.*

- *İki yöntemde de bir tuning parametresi vardır ve o parametre bileşen sayısıdır.*

- *İki yöntemde de optimum bileşen hesabında **Cross Validation** kullanılır.*
