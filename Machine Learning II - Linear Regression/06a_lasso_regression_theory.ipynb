## Lasso Regresyon (Teori)

*Amaç hata kareler toplamını minimize eden katsayıları bu katsayılara ceza işlemi uygulayarak bulmaktır. Ridge Regresyon ile neredeyse aynıdır. Fakat burada ceza uygulama işlemi biraz daha abartıldır. Katsayıların cezalarını, onları sıfır yapacak şekilde uygular. Böylece değişken seçimi yaparız.*

### Özellikler

- *Ridge Regresyon'un ilgili - ilgisiz tüm değişkenleri modelde bırakma dezavantajını gidermek için önerilmiştir.*<br>
- *Lasso'da katsayılar sıfıra yaklaştırılır.*<br>
- *Fakat L1 normu lambda değeri yeteri kadar büyük olduğunda bazı katsayıları sıfır yapar. Böylece değişken seçimi yapmış olur.*<br>
- *Lambda değerinin doğru seçilmesi çok önemlidir, burada da CV kullanılır.*<br>
- *Ridge ve Lasso yöntemleri birbirinden üstün değildir.*

## Lambda(λ) Ayar Parametresinin Belirlenmesi

- *Lambda değerinin sıfır oluğu yer EKK'dir. HKT değerini minimum yapan lambda değerini arıyoruz.*<br>
- *λ için belirli değerleri içeren bir küme seçilir ve her bir küme için cross validation test hatası hesaplanır.*<br>
- *En küçük Cross Validation'ı veren λ ayar parametresi olarak seçilir.*<br>
- *Son olarak seçilen bu λ ile model tüm gözlemlere fit edilir.*
