### Doğrusal Regresyon Varsayımları

- *Hatalar normal dağılır.*<br>
- *Hatalar birbirinden bağımsızdır ve otokorelasyon yoktur.*<br>
- *Her bir gözlem için hata terimleri varyansları sabittir.*<br>
- *Değişkenler ile hata terimi arasında ilişki yoktur.*<br>
- *Bağımsız değişkenler arasında çoklu doğrusal ilişki yoktur.*

### Regresyon Modellerinin Avantaj ve Dezavantajları

- *İyi anlaşılırsa diğer ML ve DL konuları çok rahat kavranır.*<br>
- *Doğrusallık nedensellik yorumları yapılabilmesini sağlar, bu durum aksiyoner ve stratejik modelleme imkanı verir.*<br>
- *Değişkenlerin etki düzeyleri ve anlamlılıkları değerlendirilebilir.*<br>
- *Bağımlı değişkendeki değişkenliğin açıklanma başarısı ölçülebilir.*<br>
- *Model anlamlılığı değerlendirilebilir.*<br>
- *Varsayımları vardır.*<br>
- *Aykırı gözlemlere duyarlıdır.*
