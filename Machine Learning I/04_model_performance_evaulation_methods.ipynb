*x1,x2 ve x3 değişkenlerini kullanarak bir regresyon modeli geliştirdiğimizi düşünelim. Bu modele değişkenleri parametre olarak tekrar girdiğimizde y_sapka tahmini değişkeni oluşturduk. y değişkeni ise gerçek değerler. Her gözlem biriminde bu gerçek değer ile tahmini değer farkını aldık. Oluşan fark değeri her bir gözlem için hata değerini ifade ediyor. Farkın negatif veya pozitif olması önemli değil. Farkın negatif veya pozitif olması aynı şey. İşaret sadece yönlü uzaklığı temsil ediyor. Bu yüzden aradaki farkların karesini alıyoruz. Hepsinin ortalamasını aldığımız zaman ortalama hata değerini buluyoruz. Örneğin bir gözlem biriminde gerçek değer 7.00 olsun. Benim eğittiğim modelin yapacağı tahmin gerçek değerden en fazla +-0.37'lik bir sapma miktarına sahip olur.*

*Bir sınıflandırma probleminde;*<br>
*Sınıf değerimiz gerçekte 1 iken 1 olarak tahmin ediyorsak True Pozitif*<br>
*Sınıf değerimiz 1 iken 0 olarak tahmin ediyorsak True Negatif*<br>
*Sınıf değerimiz 0 iken 1 olarak tahmin ediyorsak False Pozitif*<br>
*Sınıf değerimiz 0 iken 0 olarak tahmin ediyorsak False Negatif*

*Daha yüksek kesinlik ve anma değerlerine sahip modeller daha başarılı kabul edilir.*
