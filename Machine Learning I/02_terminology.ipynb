## Makine Öğrenmesinde Temel Kavramlar

*Temel kavramların neyi ifade ettiğini kafamızda oturtmak çok önemli çünkü makine öğrenmesinde her şey çok soyut. Kafada canlandırmak bazen zor olabiliyor.*

### Bağımlı Değişken & Bağımsız Değişken

*Yukarıda gördüğümüz y1 değişkeni bağımlı, x1,x2,x3,x4 değişkenleri bağımsız değişkenlerdir. Bağımlı değişken, bağımsız değişkenlerin etkileşimleri sonucu ortaya çıkardıkları durumun ne olduğu bilgisini taşır. Yanıt değişkeni veya çıktı değişkeni isimleriyle de görebilirsiniz.*

### Gözetimli Öğrenme | Gözetimsiz Öğrenme | Yarı Gözetimli Öğrenme

*Gözetimli Öğrenmeden dolaylı olarak yukarıda bahsettik. Bağımsız değişkenlerin etkileşimi sonucu ortaya çıkan durum veri setinde görülebiliyorsa bu gözetimli öğrenme oluyor.*

*Veri setindeki gözlemler incelenip bir yorumlama sonucu ortaya çıkan anlamlı veriler veri setinde görünmüyorsa bu gözetimsiz öğrenme oluyor. Yarı gözetimli öğrenme ise diğer iki öğrenme özelliklerini de barındıran öğrenme biçimidir.*

### Regresyon & Sınıflandırma Problemleri

*Problemimiz bir regresyon problemi ise bağımlı değişkenimiz süreklidir(float). Eğer bir sınıflandırma problemi ile uğraşıyorsak bağımlı değişkenimiz (1 ve 0, kadın ve erkek) kesikli(int) değişken oluyor.*

### Değişken Türleri

##### Sayısal Değişkenler(nicel, kantitatif)
##### Kategorik Değişkenler(nitel,kalitatif)

### Ölçek Türleri

##### Sayısal Değişkenler için : Aralık ve Oran
##### Kategorik Değişkenler için : Nominal ve Ordinal

### Test-Train Ayrımı

*Ön işlemeden çıkarılmış veri setininin %80'ini eğitim seti %20'sini test seti olarak ikiye ayırırız. Eğitim seti ile modeli eğitiriz, test seti ile eğittiğimiz modelin performans metriklerini ölçeriz. Model eğitimi sırasında eğittimiz algoritma doğal olarak test setini görmez.*

### Değişken Mühendisliği(Feature Engineering)

*Proje üzerinde çalışan kişinin veya otonom bir sistemin veri seti oluşturma aşamasında değişkenler oluşturma işlemidir. Standartlaştırma, değişken dönüşümü, normalizasyon, 0-1 dönüşümü, 1 ve diğerleri dönüşümü gibi işlemler bu adımda yapılır*

### Değişken Seçimi(Variable Selection)

*Yukarıda renkli oklarla gösterilen değişkenler, Y bağımlı değişkenini tahmin etmek konusunda başarılı olan, ayırt ediciliği olan değişkenlerdir. Kesikli çizgi ile gösterilen değişkenler ise bu ayırt edici özelliğe sahip değildir, hedef değişken tahmini konusunda başarısız olan ve kullanıcı tarafından devre dışı bırakılan değişkenlerdir.*

### Model Seçimi

#### Model seçimi için iki durum söz konusu:

***Birincisi**: Oluşabilecek değişken kombinasyonları ile oluşturulan modeller arasından en iyi modelin seçilmesi.*<br>
***İkincisi**: Kurulan birbirinden farklı modeller arasından model seçimi.*


##### Model Neye Göre Seçilir ?

*Regresyon için açıklanabilirlik oranı ve RMSE(Ortalama kare hata oranı) benzeri bir değer.*

*Sınıflandırma için doğru sınıflandırma oranı benzeri bir değer.*

### Aşırı Öğrenme(Overfitting)

*Basitçe modelin eğitim verisini aşırı öğrenmesi durumudur. Bir nevi eğitim veri setini zorla öğreniyor,ezberliyor. Ezberci bir zihniyete sahip oluyor. Öğrendiği veri seti üzzerindeki performansı iyidir fakat benzer veri setlerinde düşük performans sergiler. Model rassallığını tamamen kaybeder.*

### Deterministik Modeller & Skotastik Modeller

*Değişkenler arasında kesin bir ilişki olduğunu varsayan modeller **Deterministik Model** adını alır. X'in alacağı her bir değere karşılık bir y değeri vardır.* <br>

*Fakat veri bilimciler daha çok tesadüfi hata barındıran **Skotastik Modeller** ile ilgilenir.*

### Doğrusal Modeller vs Doğrusal Olmayan Modeller

*Doğrusal ve doğrusal olmayan model kavramlarının bilincinde olmak bu modellerin yorumlanması konusunda bize farkındalık sağlayacaktır.*

*Ağırlıklı olarak doğrusal olmayan modeller üzerinde duracağız.*
