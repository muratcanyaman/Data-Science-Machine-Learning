## Model Doğrulama Yöntemleri

*Modellerin ürettiği sonuçların doğru değerlendirilmesi çalışmalarıdır.*

*En sık kullanılan yöntemlerden bahsedeceğiz.*

### Holdout Yöntemi

    <img src="image III.png" width="400">
</div>

*Veri seti eğitim ve test seti olarak ikiye ayrılır. Eğitim seti ile eğitilen modelin başarısı test seti ile kontrol edilir. Veri setindeki gözlem sayısı az ise pek tercih edilmez. Çünkü orijinal veri setini parçalayacağımız zaman %80'in ne kadarını eğitim setine, %20'nin ne kadarını test setine ayıracağız problemi ortaya çıkıyor.*

### K-Katlı Çapraz Doğrulama(k-fold cross validation)

<div align="center">
    <img src="image VII.png" width="400">
</div>

*Eğitim seti k adet alt kümeye ayrılır. Alt kümelerden biri dışarıda bırakılır. Diğer kümelerle model eğitilir. Dışarıda kalan küme ile model test edilir. Yukarıdaki resimde veri seti 5 alt kümeye ayrılmış. 1 numaralı alt parça dışarda bırakılır, kalan 4 parça ile model eğitilir ve 1 numaralı alt küme ile modelin başarısı test edilir. Bu işlem her bir alt kümenin devre dışı bırakılıp daha sonra bu alt küme ile modelin test edilmesi tamamlanana kadar devam eder.*

*Bunu yapmamızdaki temel sebep şu : Eğitim hatası, test hatasının kötü (iyimser) bir tahmincisidir. Modelin gerçek performansını daha güvenilir bir şekilde tahmin edebilmek için eğitim setini K adet alt parçaya (fold) ayırırız. Her iterasyonda bir parça test seti, geri kalanlar eğitim seti olarak kullanılır. Elde edilen K adet hata değerinin ortalaması alınarak performans tahmini yapılır. Bu sayede tek bir train-test split'e bağlı kalmaktan kaynaklanan rastgelelik (varyans) azaltılır ve modelin genelleme performansı hakkında daha güvenilir bir tahmin elde edilir. Ayrıca bu yöntem, modelin overfit olup olmadığını tespit etmemize de yardımcı olur.*

*Yani elde edeceğimiz hatanın şansa bağlı olup olmama durumunu, rastgeleliği ortadan kaldırıyoruz. Çünkü ilk başta belki kolay örnekler teste düştü ve hata düşük çıktı, ya da zor örnekler teste düştü ve hata yüksek çıktı bunu bilemeyiz.*

### Leave One Out

<div align="center">
    <img src="image VIII.png" width="400">
</div>

*k-cross validation türevidir. Veri seti gözlem sayısı kadar alt kümeye ayrılır. n tane alt kümenin 1 tanesi test seti n-1 tanesi eğitim seti olarak kullanılır. Bu işlem ile n tane alt kümeyi test etmiş oluruz. Fakat çok fazla gözlem sayısının olduğu veri setlerinde düşük performans sergiler.*

### Yerine Koymalı Örnekleme Yöntemi

<div align="center">
    <img src="image XI.png" width="500">
</div>

*Orijinal veri setinden rastgele ve yerine koyarak (with replacement) örneklem çekilir. Yerine koyarak çekildiği için aynı gözlem birden fazla kez seçilebilir. Seçim sırasında hiç seçilmeyen gözlemler ise test seti (Out-of-Bag) olarak kullanılır. Matematiksel olarak bir gözlemin hiç seçilmeme olasılığı yaklaşık %36.8'dir, yani her iterasyonda verinin kabaca %63.2'si eğitim, %36.8'i test olarak ayrılmış olur. Bu işlem istenilen sayıda tekrarlanır ve elde edilen hata değerlerinin ortalaması alınarak modelin performansı tahmin edilir.*
