## Model Performansını Artırma

### Parametre Nedir?

*Parametre, modelin eğitim süreci boyunca veriden kendi kendine öğrendiği değerlerdir. Bunları biz belirlemeyiz; model, kayıp fonksiyonunu minimize etmeye çalışırken bu değerleri otomatik olarak bulur.*

*Örneğin doğrusal regresyondaki katsayılar ve sabit terim, ya da yapay sinir ağlarındaki ağırlıklar birer parametredir. Kısaca parametreler, modelin veriden öğrendiği şeylerdir. Eğitim tamamlandıktan sonra modelin içine işlemiş haldedirler.*

### Hiperparametre Nedir?

*Hiperparametre, modelin eğitiminden önce biz tarafından belirlenen ve modelin nasıl öğreneceğini kontrol eden değerlerdir. Model bunları veriden öğrenemez.*

*Örneğin bir karar ağacındaki maksimum derinlik değeri, Random Forest'taki ağaç sayısı, KNN'deki komşu sayısı ya da sinir ağlarındaki öğrenme hızı birer hiperparametredir. Hiperparametreler, modeli kurarken yaptığımız tasarım tercihlerdir. Farklı değerler seçmek modelin öğrenme biçimini doğrudan etkiler.*

### Parametre ile Hiperparametre Arasındaki Fark

| | **Parametre** | **Hiperparametre** |
|---|---|---|
| **Kim belirler?** | Model (otomatik olarak) | Biz (elle) |
| **Ne zaman belirlenir?** | Eğitim sırasında | Eğitimden önce |
| **Amacı** | Veriyi en iyi açıklamak | Modelin nasıl öğreneceğini kontrol etmek |
| **Örnekler** | Katsayılar, ağırlıklar | max_depth, k, learning_rate |

### Model Performansı ile İlişkisi

*Yanlış hiperparametre seçimi doğrudan underfitting veya overfitting problemine yol açar.*

*Örneğin bir karar ağacında maksimum derinlik değerini çok küçük tutarsak model yeterince öğrenemez ve underfitting yaşanır. Bu değeri çok büyük tutarsak model eğitim verisini ezberler, yeni veriye genelleyemez ve overfitting yaşanır. Doğru değer, bu ikisi arasındaki denge noktasıdır.*

*Bu yüzden hiperparametreler, modelin başarısını belirleyen en kritik tasarım kararlarından biridir. İyi bir model sadece doğru algoritmayı seçmekle değil, o algoritmanın hiperparametrelerini doğru ayarlamakla kurulur.*

### Model Tuning Nedir?

*Model tuning, bir model için en iyi performansı veren hiperparametre değerlerini sistematik olarak arama sürecidir. Bunu elle deneme-yanılma yöntemiyle yapmak mümkün olsa da bu yaklaşım hem zaman alıcıdır hem de güvenilir değildir. Bu yüzden otomatik yöntemlere başvurulur.*

#### Grid Search

*Belirlenen hiperparametre değerlerinin tüm kombinasyonlarını tek tek dener. Arama kapsamlıdır fakat parametre sayısı arttıkça hesaplama maliyeti katlanarak büyür.*

#### Randomized Search

*Belirlenen parametre uzayından rastgele kombinasyonlar örnekler ve bunları değerlendirir. Her kombinasyonu denemez fakat çok daha hızlıdır. Geniş parametre uzaylarında tercih edilir.*

*Her iki yöntemde de cross-validation kullanılır. Yani her kombinasyonun performansı, eğitim verisinin farklı bölümleri üzerinde test edilerek daha güvenilir bir şekilde ölçülür.*

### Önemli Kural

*Hiperparametre seçimi test seti üzerinden yapılmamalıdır. Test seti yalnızca en sonunda nihai modeli değerlendirmek için bir kez kullanılır. Tuning süreci her zaman eğitim seti ve cross-validation üzerinde yürütülür. Aksi takdirde test setini sonuçsuz bir şekilde eğitim sürecine dahil etmiş oluruz ve elde ettiğimiz performans değerleri gerçekçi olmaz.*
