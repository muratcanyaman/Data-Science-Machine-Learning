## Yanlılık - Varyans Değiş Tokuşu

### Modellerin Tahmin Başarılarının Değerlendirilmesi

*Modeli doğruladık. Sonrasında modelin başarısını değerlendirdik. Şimdi de bu başarıların değerlendirilmesini yapacağız. Eğitim hatası ile test hatası arasındaki ilişkiyi değerlendireceğiz.*

</div>

*Yukarıdaki 3 grafikte de çizgiler tahmini değerleri, noktalar ise gerçek değerleri temsil ediyor. Underfitting grafiğinde tahmini değer çizgisi gerçek değerleri hep es geçmiş, yanlamış. Çizgi gerçek değerlerden uzak kalmış. Fonksiyonumuz gerçek değerleri tahmin etmede yüksek oranda başarısız olmuş. Dikkat edilirse sadece belirli değerlerin üzerinden geçmiş. Buna yüksek yanlılık diyoruz. Esneklik düşük.*

*Overfitting grafiğinde tahmini değer çizgimiz bütün gerçek değerlerin üzerinden geçmiş. Veri setinin tamamını çok iyi öğrenmiş, gereğinden fazla bir şekilde. Fonksiyonumuz burada aşırı esnek yani varyansı yüksek. Çok yüksek varyanslı bir fonksiyon olmasını istemeyiz. Yazdığımız model kendisini veri setine özelleştirmiş. Yani sadece bu veri setinden çok iyi öğrenmiş de diyebiliriz. Ezber yapmış. Halbuki biz genellenebilir modeller eğitmeyi isteriz. Yani benzer veri setlerinde benzer yüksek performanslar veren modeller oluşturmak isteriz.*

*Tıpkı doğru model grafiğinde olduğu gibi. Bir model eğitiminde sadece yüksek başarı skoruna odaklanmak da yanlıştır. Bizim elde etmek isteyeceğimiz model; Genellenebilirlik kaygısını en iyi karşıladığını varsaydığımız esneklik yapısındaki düşük varyanslı düşük yanlı model olmalıdır.*

### Eğitim Hatası vs Test Hatası

<div align="center">
    <img src="image XIX.png" width="500">
</div>

*Grafikte Eğitim Seti Hatası ile Test Seti Hatası belirli bir noktaya kadar azalıyor. Modeli zorluyoruz, daha da karmaşıklaştırıyoruz, daha esnek hale getiriyoruz, parametre ekliyoruz vs fakat eğitim hatası ile test hatası arasındaki uçurum gittikçe artıyor. Model öğrenme kabiliyetini yavaş yavaş kaybediyor. Model verinin yapısını değil verinin kendisini öğreniyor. Yani veriyi birebir mapliyor, ezberliyor.*

*Modelin en optimum halini bulmanın yolu: Grafikteki gibi eğitim ve test hatası arasındaki farkın açıldığı noktayı bulmak. Modelin aşırı öğrenmeye başladığı yeri tespit etmekten geçiyor. Bu da defalarca deneme-yanılma,model eğitimi yapmaktan geçiyor.*
