## Makine Öğrenmesi Nedir ?

*Bilgisayarların insanlara benzer şekilde öğrenmesini sağlamak maksadıyla çeşitli algoritma ve tekniklerin geliştirilmesi için çalışılan bilim dalıdır.*<br>

*Daha teknik bir ifadeyle; makine öğrenmesi, bilgisayarların açıkça programlanmadan veriden öğrenmesini ve bu veriden elde ettiği örüntülere dayanarak tahminlerde bulunmasını ya da kararlar almasını sağlayan bir yapay zeka alt dalıdır.*

*Geleneksel programlamada, programcı kuralları açıkça tanımlar ve bilgisayar bu kurallara göre çıktı üretir. Makine öğrenmesinde ise bilgisayara kurallar yerine veri ve beklenen çıktılar verilir; bilgisayar bu veriden kuralları kendisi öğrenir.*

| | **Geleneksel Programlama** | **Makine Öğrenmesi** |
|---|---|---|
| **Girdi** | Veri + Kurallar | Veri + Beklenen Çıktılar |
| **Çıktı** | Sonuçlar | Kurallar (Model) |
| **Yaklaşım** | İnsan kuralları yazar | Bilgisayar kuralları öğrenir |

*Arthur Samuel (1959) makine öğrenmesini şu şekilde tanımlamıştır:*

> *"Bilgisayarlara açıkça programlanmadan öğrenme yeteneği veren çalışma alanıdır."*

*Tom Mitchell (1997) ise daha formal bir tanım yapmıştır:*

> *"Bir bilgisayar programı, eğer bir T görevi sınıfındaki performansı P ölçütüne göre E deneyimiyle iyileşiyorsa, o program T görevleri ve P performans ölçütüne göre E deneyiminden öğrenmiş sayılır."*

### Makine Öğrenmesi Neden Önemlidir ?

*Bazı problemler geleneksel programlama yaklaşımlarıyla çözülemeyecek kadar karmaşıktır. Örneğin, bir e-postanın spam olup olmadığını belirlemek için tüm kuralları elle yazmak neredeyse imkansızdır. Spam gönderenler sürekli yeni yöntemler geliştirdiğinden, kurallar da sürekli güncellenmek zorundadır.*<br>

*Makine öğrenmesi bu tür dinamik ve karmaşık problemlerde, veriden sürekli öğrenerek kendini güncelleyebilmesi nedeniyle büyük avantaj sağlar.*

*Makine öğrenmesine ihtiyaç duyulmasının temel nedenleri:*

- Kuralların çok karmaşık olduğu ve elle yazılamayacağı problemler
- Verilerin sürekli değiştiği ve kuralların güncellenmesi gereken durumlar
- Büyük veri kümelerinden anlamlı örüntülerin çıkarılması gerekliliği
- İnsan algısının yetersiz kaldığı yüksek boyutlu veri analizi

### Makine Öğrenmesi Türleri

*Makine öğrenmesi genel olarak üç ana kategoriye ayrılır:*

- **Denetimli Öğrenme (Supervised Learning)**
  - Etiketlenmiş (labeled) veri ile çalışır
  - Giriş-çıkış eşlemeleri öğrenilir
  - Regresyon ve sınıflandırma problemlerini kapsar
  - Örnek: Ev fiyat tahmini, e-posta spam tespiti

- **Denetimsiz Öğrenme (Unsupervised Learning)**
  - Etiketsiz (unlabeled) veri ile çalışır
  - Verideki gizli yapılar ve örüntüler keşfedilir
  - Kümeleme ve boyut indirgeme problemlerini kapsar
  - Örnek: Müşteri segmentasyonu, anomali tespiti

- **Pekiştirmeli Öğrenme (Reinforcement Learning)**
  - Bir ajan çevreyle etkileşime girerek öğrenir
  - Ödül ve ceza mekanizması ile doğru davranışı keşfeder
  - Örnek: Oyun oynayan yapay zeka, otonom araç sürüşü

### Gerçek Hayatta Makine Öğrenmesi Kullanım Alanları

#### Sağlık

*Makine öğrenmesi sağlık sektöründe hastalık teşhisi, ilaç keşfi ve kişiselleştirilmiş tedavi planlamalarında yaygın olarak kullanılmaktadır.*

- **Tıbbi görüntü analizi:** Röntgen, MR ve BT taramalarından tümör, kırık veya anormal yapıların otomatik tespiti. Örneğin, Google DeepMind'ın geliştirdiği model göz hastalıklarını doktorlar kadar doğru teşhis edebilmektedir.
- **Hastalık riski tahmini:** Hasta geçmişi ve genetik veriler kullanılarak diyabet, kalp hastalığı gibi kronik hastalıkların erken aşamada tahmin edilmesi.
- **İlaç geliştirme:** Yeni ilaç adaylarının keşfi ve klinik deneme süreçlerinin hızlandırılması.

#### Finans

*Finans sektörü, makine öğrenmesinin en erken ve en yoğun benimsendiği alanlardan biridir.*

- **Dolandırıcılık tespiti:** Kredi kartı işlemlerinde anormal harcama kalıplarının gerçek zamanlı tespiti. Bankalar, müşterilerinin alışveriş alışkanlıklarını öğrenerek şüpheli işlemleri anında işaretleyebilmektedir.
- **Kredi skorlama:** Bireylerin kredi ödeme güçlerinin geçmiş finansal verilerine dayanarak tahmin edilmesi.
- **Algoritmik ticaret:** Piyasa verilerinin analiz edilerek otomatik alım-satım kararlarının verilmesi.
- **Risk yönetimi:** Portföy risklerinin tahmin edilmesi ve optimize edilmesi.

#### E-ticaret ve Pazarlama

*E-ticaret platformları makine öğrenmesini müşteri deneyimini kişiselleştirmek ve satışları artırmak için kullanmaktadır.*

- **Öneri sistemleri:** Amazon, Netflix ve Spotify gibi platformlar kullanıcıların geçmiş tercihlerine dayanarak ürün, film ya da müzik önerilerinde bulunur. Netflix'in izleme önerilerinin yaklaşık %80'i öneri algoritmaları tarafından oluşturulmaktadır.
- **Müşteri segmentasyonu:** Müşterilerin satın alma davranışlarına göre gruplara ayrılması ve her gruba özel pazarlama stratejilerinin oluşturulması.
- **Dinamik fiyatlandırma:** Talep, rekabet ve kullanıcı davranışlarına göre fiyatların otomatik olarak ayarlanması.
- **Churn tahmini:** Müşterin hizmeti terk etme olasılığının tahmin edilmesi ve önlem alınması.

#### Doğal Dil İşleme (NLP)

*İnsan dilini anlama ve işleme konusunda makine öğrenmesi büyük ilerlemeler kaydetmiştir.*

- **Dil çevirisi:** Google Translate gibi araçlar makine öğrenmesi modelleri kullanarak yüzden fazla dil arasında çeviri yapmaktadır.
- **Duygu analizi (Sentiment Analysis):** Sosyal medya paylaşımları, ürün yorumları gibi metinlerin olumlu, olumsuz veya nötr olarak sınıflandırılması.
- **Chatbot ve sanal asistanlar:** Siri, Alexa ve Google Asistan gibi sesli asistanlar kullanıcıların doğal dildeki komutlarını anlayarak yanıt üretir.
- **Metin özetleme:** Uzun belgelerin otomatik olarak kısa özetlerinin oluşturulması.

#### Otonom Araçlar ve Ulaşım

*Otonom araç teknolojisi makine öğrenmesinin en ileri uygulamalarından biridir.*

- **Otonom sürüş:** Tesla, Waymo gibi şirketler derin öğrenme modelleri kullanarak araçların çevreyi algılamasını, trafik kurallarına uymasını ve güvenli sürüş yapmasını sağlamaktadır.
- **Rota optimizasyonu:** Uber ve Lyft gibi sürüş paylaşımı uygulamaları, trafik yoğunluğu ve talep tahminlerine göre en verimli rotaları belirler.
- **Trafik yönetimi:** Trafik akışının gerçek zamanlı olarak izlenmesi ve sinyal sürelerinin optimize edilmesi.

#### Görüntü ve Video Analizi

*Bilgisayarlı görü (Computer Vision) alanı makine öğrenmesi sayesinde büyük atılımlar yapmıştır.*

- **Yüz tanıma:** Telefonların yüz kimliği ile kilit açması, havalimanlarında güvenlik kontrolleri ve fotoğraf uygulamalarında otomatik etiketleme.
- **Nesne tespiti:** Üretim hatlarında kalite kontrol, güvenlik kameralarında şüpheli nesne tespiti.
- **Görüntü sınıflandırma:** Dermatolog düzeyinde cilt hastalığı teşhisi, tarımda bitki hastalığı tespiti.

#### Üretim ve Endüstri

*Endüstriyel üretimde makine öğrenmesi verimlilik artışı ve maliyet düşürme konularında kritik bir rol üstlenmektedir.*

- **Kestirimci bakım (Predictive Maintenance):** Makine sensörlerinden toplanan verilerle ekipman arızalarının önceden tahmin edilmesi. Bu sayede plansız duruşlar önlenerek üretim sürekliliği sağlanır.
- **Kalite kontrol:** Üretim hattındaki ürünlerin görüntü analizi ile otomatik olarak kontrol edilmesi ve hatalı ürünlerin ayrılması.
- **Tedarik zinciri optimizasyonu:** Talep tahmini, stok yönetimi ve lojistik planlamanın optimize edilmesi.

#### Siber Güvenlik

*Siber güvenlik alanında makine öğrenmesi, sürekli gelişen tehditlere karşı dinamik bir savunma mekanizması oluşturmaktadır.*

- **Saldırı tespiti:** Ağ trafiğinin analiz edilerek normal dışı aktivitelerin ve olası saldırıların tespit edilmesi.
- **Zararlı yazılım analizi:** Yeni ve daha önce görülmemiş zararlı yazılımların davranış kalıplarına göre tespit edilmesi.
- **Kimlik doğrulama:** Kullanıcıların klavye kullanım alışkanlıkları, fare hareketleri gibi davranışsal biyometrik verilere dayalı kimlik doğrulama.

### Makine Öğrenmesi Genel İş Akışı

*Bir makine öğrenmesi projesinin temel adımları aşağıdaki gibidir:*

1. **Problemin Tanımlanması:** Çözülmek istenen problem net bir şekilde belirlenir.
2. **Veri Toplama:** Probleme uygun veriler toplanır.
3. **Veri Ön İşleme:** Eksik veriler, aykırı değerler temizlenir; veriler modele uygun formata dönüştürülür.
4. **Keşifsel Veri Analizi (EDA):** Verinin yapısı, dağılımı ve değişkenler arası ilişkiler incelenir.
5. **Model Seçimi:** Probleme uygun algoritma veya algoritmalar belirlenir.
6. **Model Eğitimi:** Seçilen model eğitim verisi ile eğitilir.
7. **Model Değerlendirme:** Modelin performansı test verisi üzerinde ölçülür.
8. **Hiperparametre Optimizasyonu:** Modelin performansını artırmak için parametreler ayarlanır.
9. **Modelin Dağıtımı (Deployment):** Başarılı bulunan model gerçek dünyaya aktarılır.
