### Bernoulli Dağılımı

*Başarılı-Başarısız, olumlu-olumsuz şeklindeki ikili olaylar ile ilgilenildiğinde kullanılan kesikli olasılık dağılımıdır.*

$$f(x;p) = p^x (1-p)^{1-x}, \quad x \in \{0,1\}$$
$$E(X) = p \quad \quad Var(X) = pq = p(1-p)$$

$f(x;p) = p^x (1-p)^{1-x}, \quad x \in {0,1}$ → Bernoulli olasılık kütle fonksiyonu
$E(X) = p$ → Beklenen değer
$Var(X) = pq = p(1-p)$ → Varyans

*Bernoulli gibi kesikli dağılımlarda olasılık 0 ve 1 gibi değerler alır.*

### Bernoulli Dağılımı Uygulama

$\displaystyle f(x|p)=p^x(1-p)^{1-x}\quad \text{for }x\in \{0,1\}$

$\displaystyle \operatorname{E} \left(X\right)=p$

$\operatorname{Var} (X)=pq=p(1-p)$

```python
from scipy.stats import bernoulli
```

```python
p = 0.6
```

```python
rv = bernoulli(p)
rv.pmf(k = 0)
```
