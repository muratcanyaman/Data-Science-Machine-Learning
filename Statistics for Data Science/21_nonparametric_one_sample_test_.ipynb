## Nonparametrik Tek Örneklem Testi (Wilcoxon İşaret Sıralama Testi)

Parametrik testler (t testi gibi) belirli varsayımlara dayanır — bunların başında **verinin normal dağılması** gelir. Peki ya verimiz bu varsayımı sağlamıyorsa? İşte bu durumda devreye **nonparametrik testler** giriyor.


---

### Ne İçin Kullanılır?

Tek örneklem Wilcoxon testi, bir grubun medyanının belirli bir değere eşit olup olmadığını test etmemizi sağlar. Yani t testinin yaptığı işin **normal dağılım varsayımı olmadan** yapılan versiyonudur.

$$
H_0: \text{Medyan} = M_0 \quad \quad H_1: \text{Medyan} \neq M_0
$$

---

### Hangi Durumlarda Kullanılır?

| Durum | Neden Nonparametrik? |
|-------|---------------------|
| Shapiro-Wilk testi sonucunda $p < 0.05$ | Veri normal dağılmıyor, t testi kullanamayız |
| Örneklem büyüklüğü çok küçük ($n < 30$) | Normallik varsayımı güvenilir şekilde test edilemiyor |
| Veride aşırı aykırı değerler var | Aykırı değerler ortalamayı bozar, medyan daha güvenilir |
| Veri sıralama (ordinal) ölçeğinde | Ortalama hesaplanamaz, sadece sıralama yapılabilir |

---

### Parametrik vs Nonparametrik Karşılaştırması

| Özellik | Parametrik (T Testi) | Nonparametrik (Wilcoxon) |
|---------|---------------------|--------------------------|
| Varsayım | Normal dağılım gerekir | Dağılım varsayımı yok |
| Merkezi eğilim ölçüsü | Ortalama ($\bar{x}$) | Medyan |
| Güç (Power) | Daha güçlü   | Biraz daha zayıf |
| Esneklik | Kısıtlı | Daha esnek   |
| Python'da | `scipy.stats.ttest_1samp()` | `scipy.stats.wilcoxon()` |

---

>   **Önemli Not:** Nonparametrik testler dağılım varsayımı gerektirmediği için daha **esnek**tir. Ancak eğer verimiz normal dağılıyorsa, parametrik testler daha **güçlüdür** (yani gerçek bir farkı yakalama olasılığı daha yüksektir). Bu yüzden önce normallik testi yaparız — eğer veri normalse t testini, değilse Wilcoxon'u tercih ederiz.
