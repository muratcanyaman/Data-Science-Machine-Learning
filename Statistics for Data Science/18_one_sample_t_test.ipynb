### Tek Örneklem T Testi

*Popülasyon ortalaması ile varsayımsal bir değer arasında istatistiksel olarak anlamlı bir farklılık olup olmadığını test etmek için kullanılan parametrik bir testtir. Örnek ortalamasına ilişkin test yapmak için kullanılır.*

## Tek Örneklem T Testi: Test İstatistiği

$$
t = \frac{\bar{x} - \mu}{\frac{s}{\sqrt{n}}} \quad \quad \quad z = \frac{\bar{x} - \mu}{\frac{\sigma}{\sqrt{n}}}
$$

**Değişkenler:**

| Sembol | Açıklama |
|--------|----------|
| $\bar{x}$ | Örneklem ortalaması |
| $\mu$ | Popülasyon ortalaması (test edilen değer) |
| $s$ | Örneklem standart sapması |
| $\sigma$ | Popülasyon standart sapması |
| $n$ | Örneklem büyüklüğü |

Hangi test istatistiğini kullanacağımıza şu kurallara göre karar veriyoruz:

1. Anakütle standart sapması **biliniyorsa** → **z istatistiği** kullanılır.
2. Anakütle standart sapması **bilinmiyorsa** ve $n > 30$ ise → **z istatistiği** kullanılır.
3. Anakütle standart sapması **bilinmiyorsa** ve $n < 30$ ise → **t istatistiği** kullanılır.

| Durum | $\sigma$ biliniyor mu? | Örneklem Büyüklüğü | Kullanılacak Test |
|-------|----------------------|--------------------|--------------------|
| 1 | ✅ Evet | Fark etmez | $z$ testi |
| 2 | ❌ Hayır | $n > 30$ | $z$ testi |
| 3 | ❌ Hayır | $n < 30$ | $t$ testi |

> 💡 $n > 30$ olduğunda, Merkezi Limit Teoremi gereği örneklem dağılımı normale yaklaştığı için $\sigma$'yı bilmesek bile $z$ istatistiğini kullanabiliyoruz.


## Tek Örneklem T Testi: Varsayım

T testini uygulayabilmemiz için verimizin sağlaması gereken temel varsayım:

### Normal Dağılım

Verinin (veya örneklemin) **normal dağılıma** uyması gerekmektedir.

---

> 📊 **Veri Bilimi Perspektifi:**
>
> Gerçek dünya verisiyle çalışırken verimiz çoğu zaman **tam olarak normal dağılmaz.** Bu durumda şu adımları izliyoruz:
>
> 1. **Normallik testi uygularız** → Shapiro-Wilk (`scipy.stats.shapiro`) veya Kolmogorov-Smirnov testi
> 2. **$n > 30$ ise** → Merkezi Limit Teoremi sayesinde örneklem ortalamaları normale yaklaşır, bu yüzden küçük sapmalar tolere edilebilir
> 3. **Veri normal dağılmıyorsa ve $n < 30$ ise** → Parametrik olmayan testlere (Wilcoxon gibi) yöneliriz
>
> | Durum | Normallik | Ne yaparız? |
> |-------|-----------|-------------|
> | $n > 30$, hafif sapma | ≈ Normal | T testi uygulanabilir ✅ |
> | $n > 30$, ciddi sapma | ❌ | Nonparametrik test düşünülür |
> | $n < 30$, normal | ✅ | T testi uygulanabilir ✅ |
> | $n < 30$, normal değil | ❌ | Nonparametrik test kullanılır |
>
> Kısacası, **test seçmeden önce her zaman normallik varsayımını kontrol etmemiz** gerekiyor.
