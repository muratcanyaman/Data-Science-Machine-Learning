### Olasılığa Giriş ve Olasılık Dağılımları

*Belirsizlikle ilgili yorumlamalar yaparken başvurulan tekniklerden biridir.*

*Rassal Değişken: Değerini bir deneyin sonuçlarından alan değişkendir.*

*Dağılım Nedir ? <br>
Gerçekleşen olayların veya durumların sayısal karşılıklarının ortaya çıkardıkları yapıya **Dağılım** denir.* 

*Olasılık Dağılımı Nedir ?<br>
Bir rassal olaya ait değerler ve bu değerlerin gerçekleşme olasılıklarının bir arada ifade edilmesine **Olasılık Dağılımı** denir.*

*Olasılık Fonksiyonu Nedir ? <br>
Bir değişkenin herhangi bir değeri alması olasılığını hesaplayan fonksiyona **Olasılık Fonksiyonu** denir.*
