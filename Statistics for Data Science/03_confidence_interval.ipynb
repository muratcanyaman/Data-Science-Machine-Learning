### Güven Aralığı

*Anakütle parametresinin tahmini değerini kapsayabilecek iki sayıdan oluşan bir aralık bulunması sonucu elde ettiğimiz sonuca denir. Yaptığımız tahminlerin ne kadar güvenilir olduğu ile ilgili bize bir değer sunar.*

*Örneğin bir websitesinde geçirilen süre ile ilgili bir problemimiz olsun. Ortalama geçirilen süre 180 sn diyelim. Bu bilgiyi daha güvenilir kılmak için bir güven seviyesi verilir ve birtakım formüller ile güven aralığı bulunur.*

*Mesela "%95 güven aralığında websitesinde gezinirken ortalama gezinilen süre 172-180 saniye aralığındadır" şeklinde bir çıkarımda bulunmuş olalım. Yani burada diyoruz ki 100 kişiden 95'i ortalama 172-180 saniye bu websitesinde geziniyor. En başta sadece ortalama bir değer varken şimdi %95 kesinlik oranında bu ortalamanın yoğunluluğunun en çok hangi iki uç değer arasında olabileceğini söyleyebiliyoruz. Üstelik düşük bir hata payı ile.*

### Güven Aralığı Nasıl Hesaplanır ? 

*Adım 1: n, ortalama ve standart sapmayı bul(n = 100, x = 180 ve s = 40 olsun).*<br>
*Adım 2: Güven aralığına karar ver. %95 mi %99 mu ? (Kabul görülen standart genellikle %95'tir.*<br>
*Adım 3: z tablo değerini hesapla (1,96 - 2,57).*<br>
*Adım 4 : Yukarıdaki değerleri kullanarak güven aralığını bul.*
*Güven Aralığı Formülü: $$\bar{x} \pm z \cdot \frac{S}{\sqrt{n}} = 180 \pm 1{,}96 \times \frac{40}{\sqrt{100}}$$*

*Yukarıdaki formülde çıkan sonuç 172 - 180'dir. Yani burada ne diyoruz? Websitede kalınan süre istatistiksel olarak %95'lik güvenilirlik oranı ile 172-180 saniyedir. Her 100 kişiden sadece 5 kişi bu aralık dışında zaman geçiriyor da diyebiliriz.*
