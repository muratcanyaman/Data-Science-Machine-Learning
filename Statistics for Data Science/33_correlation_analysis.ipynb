*Korelasyonların anlamlılıklarını hipotez testi ile inceleriz.*

**Hipotezler**

$$
\begin{aligned}
H_0&: \rho = 0 \\
H_1&: \rho \neq 0
\end{aligned}
$$


*Burada H0: Değişkenler arasında anlamlı bir ilişki olmadığını söyler. H1 ise anlamlı bir ilişki yani korelasyon olduğunu söyler. H0'ın reddedilmesini bekleriz çünkü değişkenler arasında anlamlı bir ilişki olmasını isteriz.*

#### Test İstatistiği

**Pearson Korelasyon Katsayısı:**

$$
r_{xy} = \frac{\sum x_i y_i - n\bar{x}\bar{y}}{\sqrt{(\sum x_i^2 - n\bar{x}^2)} \sqrt{(\sum y_i^2 - n\bar{y}^2)}}
$$

| Sembol | Açıklama |
|--------|----------|
| $r_{xy}$ | Pearson korelasyon katsayısı ($-1$ ile $+1$ arasında) |
| $x_i$, $y_i$ | İki değişkenin gözlem değerleri |
| $\bar{x}$, $\bar{y}$ | İki değişkenin ortalamaları |
| $n$ | Örneklem büyüklüğü |

Bu formül iki değişken arasındaki **doğrusal ilişkinin yönünü ve şiddetini** hesaplamamızı sağlar.

---

**Korelasyonun Anlamlılık Testi:**

$$
t = r \sqrt{\frac{n - 2}{1 - r^2}}
$$

| Sembol | Açıklama |
|--------|----------|
| $t$ | Hesaplanan test istatistiği |
| $r$ | Hesaplanan korelasyon katsayısı |
| $n$ | Örneklem büyüklüğü |

Korelasyon katsayısını hesapladıktan sonra, bu ilişkinin **istatistiksel olarak anlamlı olup olmadığını** bu t testi ile doğrularız. Yani $r$ değeri sıfırdan yeterince farklı mı, yoksa şans eseri mi ortaya çıkmış sorusuna cevap ararız.


*Pierson korelasyonu pozitifse pozitif yönlü korelasyon,  negatifse negatif yönlü korelasyon vardır deriz. Eğer sıfırsa değişkenler arasında bir ilişki olmadığı sonucunu elde ederiz.*
