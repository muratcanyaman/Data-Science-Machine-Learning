## Normal Dağılım

Normal dağıldığı bilinen sürekli rassal değişkenler için olasılık hesaplaması için kullanılır.

$$
f(x \mid \mu, \sigma^2) = \frac{1}{\sqrt{2\pi\sigma^2}} \, e^{-\frac{(x - \mu)^2}{2\sigma^2}}
$$

**Değişkenler:**

| Sembol | Açıklama |
|--------|----------|
| $x$ | Rassal değişkenin aldığı değer |
| $\mu$ | Ortalama (mean) – dağılımın merkezi |
| $\sigma$ | Standart sapma (standard deviation) – dağılımın yayılımı |
| $\sigma^2$ | Varyans – standart sapmanın karesi |
| $\pi$ | Pi sayısı ($\approx 3.14159$) |
| $e$ | Euler sayısı ($\approx 2.71828$) |

---

**Notlar:**

- Normal dağılım **çan eğrisi (bell curve)** şeklinde, simetrik bir dağılımdır.
- Eğri $\mu$ değerine göre **simetrik** olup, ortalama = medyan = mod'dur.
- **68-95-99.7 Kuralı (Empirik Kural):**
  - Verilerin **%68**'i → $\mu \pm 1\sigma$ aralığında
  - Verilerin **%95**'i → $\mu \pm 2\sigma$ aralığında
  - Verilerin **%99.7**'si → $\mu \pm 3\sigma$ aralığında
- $\mu = 0$ ve $\sigma = 1$ olduğunda **Standart Normal Dağılım (Z dağılımı)** elde edilir.
- Eğrinin altında kalan toplam alan **1**'e eşittir (toplam olasılık).
