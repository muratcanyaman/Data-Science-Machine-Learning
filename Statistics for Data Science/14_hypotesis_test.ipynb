## Hipotezler ve Türleri

Hipotez testi, elimizdeki verinin bir iddayı destekleyip desteklemediğini **istatistiksel olarak** kontrol etmemizi sağlar.

Her hipotez testinde iki hipotez vardır:

- $H_0$ **(Null Hipotez / Sıfır Hipotezi):** Mevcut durumu savunur. "Bir fark/etki yoktur" der. Biz bunu **çürütmeye** çalışırız.
- $H_1$ **(Alternatif Hipotez):** Bizim iddiamızdır. "Bir fark/etki vardır" der. Biz bunu **kanıtlamaya** çalışırız.

> 💡 Yazılımdaki gibi düşünebiliriz: $H_0$ sistemin **default davranışıdır**; $H_1$ ise bizim getirdiğimiz **yeni feature/değişiklik**. Test edip anlamlı bir fark yoksa default kalır, varsa yeni değişiklik kabul edilir.

---

### 1. İki Yönlü Test (Two-Tailed Test)

$$
H_0: \mu = 50 \quad \quad H_1: \mu \neq 50
$$

- **Ne sorar?** → Değer 50'den **farklı mı?** (büyük veya küçük fark etmez)
- **Kullanım:** Yönü belli olmayan bir farkı test etmek istediğimizde.
- **Örnek:** Bir A/B testinde yeni butonun tıklanma oranı eskisinden **farklı mı?** (daha iyi mi daha kötü mü bilmiyoruz, sadece fark var mı diye bakıyoruz)

---

### 2. Sağ Kuyruk Testi (Right-Tailed Test)

$$
H_0: \mu \leq 50 \quad \quad H_1: \mu > 50
$$

- **Ne sorar?** → Değer 50'den **büyük mü?**
- **Kullanım:** Bir artış/iyileşme olduğunu kanıtlamak istediğimizde
- **Örnek:** Yeni bir caching algoritması API response süresini **artırdı mı?** Yeni model accuracy'si eski modelden **daha yüksek mi?**

---

### 3. Sol Kuyruk Testi (Left-Tailed Test)

$$
H_0: \mu \geq 50 \quad \quad H_1: \mu < 50
$$

- **Ne sorar?** → Değer 50'den **küçük mü?**
- **Kullanım:** Bir azalma/düşüş olduğunu kanıtlamak istediğimizde
- **Örnek:** Yeni optimizasyon sayfa yüklenme süresini **düşürdü mü?** Bug fix'ten sonra hata oranı **azaldı mı?**

---

### 📋 Özet Tablo

| Test Türü | $H_0$ | $H_1$ | Soru |
|-----------|--------|--------|------|
| **İki Yönlü** | $\mu = 50$ | $\mu \neq 50$ | Farklı mı? |
| **Sağ Kuyruk** | $\mu \leq 50$ | $\mu > 50$ | Daha büyük mü? |
| **Sol Kuyruk** | $\mu \geq 50$ | $\mu < 50$ | Daha küçük mü? |

> 🔑 **Önemli:** Biz hiçbir zaman $H_0$'ı **kanıtlamayız**, sadece **reddederiz veya reddedemeyiz.** Tıpkı yazılımda bir bug'ı "yok" diye kanıtlayamayacağımız gibi — sadece "bulamadım" diyebiliriz.
