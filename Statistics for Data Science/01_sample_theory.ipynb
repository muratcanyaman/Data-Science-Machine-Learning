### Örnek Teorisi

*Bir popülasyondan rastgele(olasılıklı) çekilen örneklerin oluşturduğu popülasyon alt kümesine "Örneklem" denir. Örneklem çekmemizin temel amacı az veriden o örneklemin çekildiği popülasyona ait genel çıkarımlar yapmaktır. Bazen popülasyon o kadar kalabalıktır ki o popülasyondan veri elde etmek çok uzun sürer, bu da zaman kaybıdır.*

```python
import numpy as np
```

```python
populasyon = np.random.randint(0, 80, 1000)
```

```python
populasyon[0 : 10]
```

```python
#Örneklem oluşturma
#Her örneklem oluşturma işleminde aynı yapıya göre sonuç getirmesi maksadıyla random.seed() ekliyorum. 
np.random.seed(115)
orneklem = np.random.choice(a = populasyon, size = 100)
orneklem[0 : 10]
```

```python
orneklem.mean()
```

```python
populasyon.mean()
```

```python
#Örneklem Dağılımı
np.random.seed(10)
orneklem1 = np.random.choice(a = populasyon, size = 100)
orneklem2 = np.random.choice(a = populasyon, size = 100)
orneklem3 = np.random.choice(a = populasyon, size = 100)
orneklem4 = np.random.choice(a = populasyon, size = 100)
orneklem5 = np.random.choice(a = populasyon, size = 100)
orneklem6 = np.random.choice(a = populasyon, size = 100)
orneklem7 = np.random.choice(a = populasyon, size = 100)
orneklem8 = np.random.choice(a = populasyon, size = 100)
orneklem9 = np.random.choice(a = populasyon, size = 100)
orneklem10 = np.random.choice(a = populasyon, size = 100)
```

```python
#Bütün örneklemlerin ortalamasının ortalamasını alalım
(orneklem1.mean() + orneklem2.mean() + orneklem3.mean() + orneklem4.mean() + orneklem5.mean()
 + orneklem6.mean() + orneklem7.mean() + orneklem8.mean() + orneklem9.mean() + orneklem10.mean()) / 10
```

*10 örneklem ortalaması hemen hemen popülasyon ortalaması ile neredeyse aynı.*
