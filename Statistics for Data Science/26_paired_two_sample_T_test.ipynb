## Bağımlı İki Örneklem T Testi (Paired Sample T Test)

Bağımlı iki örneklem t testi, **aynı grup** üzerinde yapılan iki farklı ölçümün (örneğin öncesi ve sonrası) ortalamaları arasında istatistiksel olarak anlamlı bir fark olup olmadığını test etmemizi sağlar.

Bağımsız iki örneklem t testinden farkı: burada iki farklı grup değil, **aynı grubun farklı zamanlardaki ölçümleri** karşılaştırılır.

### Ne Zaman Kullanırız?

- Bir tedavi, eğitim veya müdahalenin **öncesi ve sonrası** ölçüm karşılaştırmalarında
- Aynı kişilere **farklı koşullarda** uygulanan testlerin sonuçlarını değerlendirirken
- Bir modelin veya sistemin **güncelleme öncesi ve sonrası** performansını kıyaslarken

Önemli olan, iki ölçümün **aynı birimlerden** (aynı kişiler, aynı ürünler vb.) alınmış olmasıdır.

### Hipotezler

| İki Yönlü Test | Sağ Kuyruk Testi | Sol Kuyruk Testi |
|----------------|------------------|------------------|
| $H_0: \mu_ö = \mu_s$ | $H_0: \mu_ö \leq \mu_s$ | $H_0: \mu_ö \geq \mu_s$ |
| $H_1: \mu_ö \neq \mu_s$ | $H_1: \mu_ö > \mu_s$ | $H_1: \mu_ö < \mu_s$ |

$\mu_ö$: Önceki ölçümün ortalaması, $\mu_s$: Sonraki ölçümün ortalaması

### Test İstatistiği

$$
t = \frac{\bar{x}_D - \mu_0}{\frac{S_D}{\sqrt{n}}}
$$

| Sembol | Açıklama |
|--------|----------|
| $\bar{x}_D$ | Farkların ortalaması |
| $\mu_0$ | Test edilen hipotez değeri (genellikle 0) |
| $S_D$ | Farkların standart sapması |
| $n$ | Örneklem büyüklüğü (çift sayısı) |

### Varsayımlar

- **Normallik:** Farklar normal dağılmalıdır. Shapiro-Wilk testi ile kontrol edilir.
- **Varyans Homojenliği:** İki ölçümün varyansları homojen olmalıdır.

Normallik varsayımı sağlanmıyorsa nonparametrik alternatifi olan **Wilcoxon İşaret Sıralama Testi** kullanılır.

### Bağımsız vs Bağımlı İki Örneklem T Testi

| Özellik | Bağımsız | Bağımlı |
|---------|----------|--------|
| Gruplar | Farklı bireyler | Aynı bireyler |
| Ölçüm | Tek seferlik | Öncesi-sonrası |
| Varyans kontrolü | Levene testi | Gerekli değil |
| Python fonksiyonu | `ttest_ind()` | `ttest_rel()` |
