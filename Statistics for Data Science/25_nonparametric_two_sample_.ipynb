## Nonparametrik Bağımsız İki Örneklem Testi (Mann-Whitney U Testi)

Bağımsız iki örneklem t testinde **normallik varsayımı** sağlanmıyorsa, parametrik olmayan alternatif olarak **Mann-Whitney U Testi** kullanırız.

Bu test, iki bağımsız grubun dağılımları arasında anlamlı bir fark olup olmadığını **medyan** üzerinden değerlendirir.

### Ne Zaman Kullanırız?

- Shapiro-Wilk testi sonucunda her iki grubun da veya en az birinin **normal dağılmadığı** tespit edildiyse
- Örneklem büyüklüğü çok küçükse ve normallik varsayımı güvenilir şekilde test edilemiyorsa
- Veride aşırı aykırı değerler varsa ve bu değerler ortalama üzerinden karşılaştırmayı güvenilmez kılıyorsa

Kısacası: Bağımsız iki örneklem t testinin ön koşulları sağlanmıyorsa, bu teste yöneliriz.

### Hipotezler
$$
\begin{aligned}
H_0&: M_1 = M_2 \quad \text{(İki grubun medyanları arasında fark yoktur)} \\
H_1&: M_1 \neq M_2 \quad \text{(İki grubun medyanları arasında anlamlı fark vardır)}
\end{aligned}
$$

### Parametrik ve Nonparametrik Karşılaştırması

| Özellik | Parametrik (T Testi) | Nonparametrik (Mann-Whitney U) |
|---------|---------------------|-------------------------------|
| Varsayım | Normal dağılım gerekir | Dağılım varsayımı yok |
| Karşılaştırma ölçüsü | Ortalama | Medyan / Sıra |
| Güç | Daha güçlü | Biraz daha zayıf |
| Esneklik | Kısıtlı | Daha esnek |
| Python fonksiyonu | `scipy.stats.ttest_ind()` | `scipy.stats.mannwhitneyu()` |

### Python ile Uygulama

```python
import pandas as pd
A = pd.DataFrame([30, 27, 21, 27, 29, 30, 20, 27, 35, 22, 24, 23, 25, 27, 23, 27, 23,
                   25, 21, 18, 24, 26, 33, 26, 27, 28, 19, 25, 23, 29])

B = pd.DataFrame([37, 39, 31, 31, 34, 38, 30, 36, 29, 28, 38, 28, 37, 37, 30, 32, 31, 31, 27,
                   32, 33, 33, 33, 31, 32, 33, 26, 32, 33, 29])
#aynı indexteki değerleri "A" ve "B" adlı sutünlara yan yana ekleyerek iki dataframe'i birleştiriyoruz.
A_B = pd.concat([A, B], axis=1)
A_B.columns = ["A", "B"]

A_B.head()
```

```python
from scipy.stats import mannwhitneyu

# Örnek: İki grubun verileri
# A_B["A"] ve A_B["B"] gibi iki bağımsız grup olduğunu varsayalım

test_istatistigi, pvalue = mannwhitneyu(A_B["A"].dropna(), A_B["B"].dropna())

print('Test İstatistiği = %.4f, p-değeri = %.4f' % (test_istatistigi, pvalue))
```

*p-value değerimiz alpha değeri olan 0.05'ten çok ama çok küçük.*

### Sonuç Yorumlama

| Koşul | Karar |
|-------|-------|
| $p \geq 0.05$ | $H_0$ reddedilemez. İki grup arasında anlamlı fark yoktur. |
| $p < 0.05$ | $H_0$ reddedilir. İki grup arasında istatistiksel olarak anlamlı fark vardır. |

### Önemli Not

Nonparametrik testler dağılım varsayımı gerektirmediği için daha esnektir. Ancak eğer verimiz normal dağılıyorsa, parametrik testler daha güçlüdür (yani gerçek bir farkı yakalama olasılığı daha yüksektir). Bu yüzden önce normallik testi yaparız; eğer veri normalse t testini, değilse Mann-Whitney U testini tercih ederiz.
