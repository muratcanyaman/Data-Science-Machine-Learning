### İş Uygulaması: Kullanıcı Arayüzü Deneyi (AB Testi)

*Bir websitesi arayüzünde bulunan buton rengine göre görüntülenme ve tıklanma bilgisi var. Eğer buton yeşil ise 1000 görüntülenme ve 300 tıklanma alırken, kırmızı iken 1100 görüntülenme ve 250 tıklanma alıyor. Bu bilgilere bakılarak hangi butonun seçileceğine dair bir hipotez testi yapılmak isteniyor.*

**Kırmızı Buton mu Yeşil Buton mu?**

$$
H_0: P_1 = P_2
$$

$$
H_1: P_1 \neq P_2
$$

| Yeşil Buton | Kırmızı Buton |
|:-----------:|:-------------:|
| 1000 görüntülenme | 1100 görüntülenme |
| 300 tıklama | 250 tıklama |


#### İki Örneklem Oran Testi

```python
from statsmodels.stats.proportion import proportions_ztest
import numpy as np
```

```python
basari_sayisi = np.array([300, 250])
gozlem_sayilari = np.array([1000, 1100])
```

```python
proportions_ztest(count = basari_sayisi, nobs = gozlem_sayilari)
```

*p - value değeri 0.05'ten çok küçük. H0'ı reddettik. H0 ne diyordu ? Birinci butonun geri dönüş oranı ile ikinci butonun dönüş oranı arasında anlamlı bir fark yoktur. Bunu reddetmiş olduk.*

*n > 30 kontrolünü yapmadık çünkü örnekte bariz olarak örneklem sayısının 30' un çok yukarısında olduğunu görüyoruz.*
