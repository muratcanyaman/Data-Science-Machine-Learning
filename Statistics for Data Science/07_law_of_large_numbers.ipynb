### Büyük Sayılar Yasası

*Bir rassal değişkenin uzun vadeli kararlılığını tanımlayan olasılık teoremidir.*

```python
import numpy as np
#RandomState(),kodu kaç kere çalıştırırsak çalıştıralım her oluşturduğumuzda dizilerin aynı kalması için kullanılır.
rng = np.random.RandomState(123)
for i in np.arange(0, 21):
    deney_sayisi = 2**i
    yazi_turalar = rng.randint(0, 2, size = deney_sayisi)
    yazi_olasiliklari = np.mean(yazi_turalar)
    print("Atış Sayısı:", deney_sayisi, "----", "Yazı Olasılığı: %.2f" % (yazi_olasiliklari * 100))
```

*Yukarıdaki çıktıdan Büyük Sayılar Yasası'nın bize tam olarak ne söylediğini açıkça anlayabiliriz : Bir deneyi ne kadar çok tekrarlarsak,gözlemlenen ortalama, beklenen değere(teorik değer) o kadar yaklaşır.Buradan beklediğimiz değer ise atılan bir paranın yazı gelme olasılığı ki bu da teorik olarak %50'dir.*<br>

*az deney: Sonuçlar rastgele ve dengesiz olabilir.*<br>
*çok deney: Sonuçlar teorik olasılığa yaklaşır.*
