### Bağımsız İki Örneklem Testi(AB Testi)

*İki grup ortalaması arasında karşılaştırma yapılmak istenildiğinde kullanılır.*<br>

*Bunun daha karmaşık tanımı şudur: Elimizde gerçek değerini bilmediğimiz iki tane ana kitle parametresi var. Ki bunlar 2 tane farklı ana kitlenin ortalamaları ve bu ortalamaların birbirinden farkını inceliyoruz.*<br>

*Buna yönelik olarak elde ettiğimiz örneklemler üzerinden karşılaştırma yapmamız gerekiyor.*<br>

*İşte biz bunları iki grup ortalamasına indirgemiş oluyoruz. Yani elimizde iki tane örnek ortalaması olacak.*<br>

*Bunların arasındaki farkı test edip, gerçekte bilmediğimiz ana kitlenin değerlerine ilişkin böyle bir farkın istatistiki olarak var olup olmadığını değerlendirmiş olacağız.*

## Hipotezler

| İki Yönlü Test | Sağ Kuyruk Testi | Sol Kuyruk Testi |
|----------------|------------------|------------------|
| $H_0: \mu_1 = \mu_2$ | $H_0: \mu_1 \leq \mu_2$ | $H_0: \mu_1 \geq \mu_2$ |
| $H_1: \mu_1 \neq \mu_2$ | $H_1: \mu_1 > \mu_2$ | $H_1: \mu_1 < \mu_2$ |


### Test İstatistiği

---

### 1. Örnek sayıları aynı, varyanslar homojen ise:

$$
t = \frac{\bar{X}_1 - \bar{X}_2}{S_p \sqrt{\frac{2}{n}}} \quad , \quad S_p = \sqrt{\frac{s_1^2 x_1 + s_2^2 x_2}{2}}
$$

---

### 2. Örnek sayısı farklı, varyanslar homojen ise:

$$
t = \frac{\bar{X}_1 - \bar{X}_2}{S_p \sqrt{\frac{1}{n_1} + \frac{1}{n_2}}} \quad \quad S_p = \sqrt{\frac{(n_1 - 1)s_1^2 + (n_2 - 1)s_2^2}{n_1 + n_2 - 2}}
$$

---

### 3. Örnek sayıları farklı, varyanslar homojen değil ise (Welch's T Testi):

$$
t = \frac{\bar{X}_1 - \bar{X}_2}{S_{\Delta}} \quad , \quad S_{\Delta} = \sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}}
$$

> 💡 Bu son formülasyon **Welch's T Testi** (Welch T Testi) olarak da bilinir. Varyansların eşit olmadığı durumlarda klasik t testine göre daha güvenilir sonuçlar verir.

---

### Parametreler

| Sembol | Açıklama |
|--------|----------|
| $\bar{X}_1$, $\bar{X}_2$ | 1. ve 2. grubun örneklem ortalamaları |
| $s_1^2$, $s_2^2$ | 1. ve 2. grubun örneklem varyansları |
| $n_1$, $n_2$ | 1. ve 2. grubun örneklem büyüklükleri |
| $S_p$ | Havuzlanmış (pooled) standart sapma — homojen varyans varsayımında kullanılır |
| $S_{\Delta}$ | Welch düzeltmeli standart hata — homojen olmayan varyanslarda kullanılır |


##  Varsayımlar

Bağımsız iki örneklem t testini uygulayabilmemiz için iki temel varsayımın sağlanması gerekir:

---

### 1. Normallik

Her iki grubun verisi de **normal dağılıma** uymalıdır. Bunu kontrol etmek için Shapiro-Wilk testini kullanırız:

- $p \geq 0.05$ → Veri normal dağılıyor  → T testine devam edebiliriz
- $p < 0.05$ → Veri normal dağılmıyor  → Nonparametrik teste (Mann-Whitney U) yönelmemiz gerekir

---

### 2. Varyans Homojenliği

İki grubun varyanslarının **birbirine yakın (homojen)** olması gerekir. Bunu kontrol etmek için Levene testini kullanırız:

- $p \geq 0.05$ → Varyanslar homojen  → Klasik t testi formülünü kullanırız
- $p < 0.05$ → Varyanslar homojen değil  → **Welch's T Testi** formülünü kullanmamız gerekir

---

*Özetle: Önce **normallik**, sonra **varyans homojenliği** kontrol edilir. Bu iki varsayımın sonucuna göre hangi testi ve hangi formülü kullanacağımıza karar veririz.*
