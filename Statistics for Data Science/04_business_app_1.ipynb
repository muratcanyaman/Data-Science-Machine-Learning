### İş Uygulaması: Fiyat Stratejisi Karar Destek Sistemi

*Problem: Bir şirket CEO'su fiyat belirleme konusunda bilimsel bir dayanak ve esneklik istiyor.*<br>
*Satıcı, alıcı ve alınan ürün var.*<br>
*Alıcılara ürüne ne kadar fiyat öderdiniz diye soruluyor.*<br>
*Optimum fiyat bilimsel ve esnek olarak bulunmak istiyor.Öyle bir fiyat belirlenecek ki tatmin edici bir gelir elde edilsin.*<br>
*Burada önemli bir detay var.Hiçbir alıcı başka bir alıcının teklif ettiği fiyatı göremiyor.*

*Şirkette çalışan bir veri bilimci olarak müşterilerden ürünü satın almak istedikleri fiyatları toplayıp ortalama değerini şirket CEO'suna direkt olarak sunabiliriz.Bilimsel olarak sunulmayan bu değer şirket satışları için ne kadar değerlidir ? Ne kadar kar getirebilir ? Bilimsel bir temele oturtularak yapılan istatistiki tahmin çok daha güçlüdür.Yüksek ihtimalle ne olacağı bellidir. Güven aralığı işte bu yüzden var. Güven aralığı ile müşteriler en çok hangi aralıkta teklif sunmuşsa ona göre bir satış fiyatının belirlenmesi şirkete daha çok kazanç getirebilir.*

*1000 tane müşteriden bütün fiyat tekliflerini bir excel dosyasından çekip listelediğimizi düşünelim. Akılda kalması için burada kendi random fiyat dizimizi oluşturacağız.*

```python
import numpy as np
fiyatlar = np.random.randint(10, 100, 1000)
```

```python
fiyatlar.mean()
```

```python
import statsmodels.stats.api as sms
```

```python
sms.DescrStatsW(fiyatlar).tconfint_mean()
```
