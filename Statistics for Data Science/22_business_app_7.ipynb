- **Problem:**
  Bir yazılım ile bir mecrada reklam verilmiş ve bu reklama ilişkin yazılım tarafından 0.125 dönüşüm oranı elde edildiği ifade edilmiş. Fakat bu durum kontrol edilmek isteniyor. Çünkü bu yüksek bir oran ve gelirler incelendiğinde örtüşmüyor.

- **Detaylar:**
  - 500 kişi dış mecrada reklamlara tıklamış, 40 tanesi sitemize gelip alışveriş yapmış.
  - Örnek üzerinden elde edilen dönüşüm oranı: $40 / 500 = 0{.}08$


*Detaylar kısmında problem tanımımız biraz değişiyor. Tıklayıp gelenler var, bir de alışveriş yapanlar var. Eğer reklamı görüp tıklayanlara odaklanırsak, görenler ve tıklayanlar odağında **rekama tıklayanlar / reklamı görenler** şeklinde bir dönüşüm oranımız olacaktır. Eğer alışveriş yapanlara odaklanırsak; **alışveriş yapanlar / reklama tıklayanlar** şeklinde bir dönüşüm oranımız olacaktır. Burada 0.08 dönüşüm oranı 2. dönüşüm oranına göre elde edilen bir veri. O yüzden 2.ci'sine odaklanacağız.*

$$
H_0: P = 0.125
$$

$$
H_1: P \neq 0.125
$$


```python
# Tek Örneklem Oran Testi
from statsmodels.stats.proportion import proportions_ztest
```

```python
count = 40
nobs = 500
value = 0.125
```

```python
proportions_ztest(count, nobs, value)
```

*p-value değerimiz kabul edilebilir hata oranı olan **0.05**'ten küçük. H0 hipotezini reddektik. Yani %95 güven düzeyi ile test ettiğimiz 0.125 dönüşüm oranı geçerliliğini kaybediyor. Gerçekten gelirler ve dönüşüm oranının örtüşmediğini bilimsel olarak kanıtlamış olduk.*
