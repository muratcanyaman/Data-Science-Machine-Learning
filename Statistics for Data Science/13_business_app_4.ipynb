### Kodlama

```python
from scipy.stats import norm
```

```python
#90'dan fazla olma olasılığı
1 - norm.cdf(90, 80, 5)
```

*Neden cdf()?<br>
Artık sürekli değişkenlerle uğraşıyoruz.Alan taraması yapacağız,kümülatif olarak değerlendirilir.Bu yüzden cdf(Kümülatif yoğunluk fonksiyonu) kullanırız.*

*Peki neden 1-norm?<br>
cdf() fonksiyonu her zaman soldan başlayarak biriktirir.Yani aslında cdf(90, 80, 5) yazdığımızda X <= 90 olan alanı tarar.Biz burada sağ kuyruğu istiyoruz(X >= 90).O yüzden 1'den çıkarma işlemi gerçekleştiririz.*

```python
#70'ten fazla olma olasılığı
1 - norm.cdf(70, 80, 5)
```

```python
#73'ten az olma olasılığı
norm.cdf(73, 80, 5)
```

```python
#84 ile 90 arasında olma olasılığı
norm.cdf(90,80,5) - norm.cdf(85,80,5)
```
