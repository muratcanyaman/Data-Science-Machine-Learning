## Hata Tipleri

Hipotez testinde bir karar verdikten sonra bu kararın **doğru mu yanlış mı** olduğunu bilmemiz gerekiyor. 4 olası sonuç vardır:

|  | $H_0$ **reddedilmedi** | $H_0$ **reddedildi** |
|--|------------------------|---------------------|
| **Gerçekte $H_0$ doğru** |    Doğru Karar · $(1 - \alpha)$ →    Güven Düzeyi |    **I. Tip Hata** · $\alpha$ |
| **Gerçekte $H_0$ yanlış** |    **II. Tip Hata** · $\beta$ |     Doğru Karar · $(1 - \beta)$ → Testin Gücü |

---

###   Doğru Kararlar

- **Güven Düzeyi $(1 - \alpha)$:** $H_0$ gerçekten doğruyken onu reddetmeme olasılığı. Genelde **%95** olarak belirlenir. Yani doğru olan bir şeyi doğru kabul etmişiz — sorun yok.

- **Testin Gücü $(1 - \beta)$:** $H_0$ gerçekten yanlışken onu reddetme olasılığı. Yanlış olan bir şeyi yakalayıp reddetmişiz — yine sorun yok.

---

###   Hatalar

- **I. Tip Hata (False Positive) — $\alpha$:**
  $H_0$ aslında doğruyken yanlışlıkla **reddettik.**
  > 🖥️ Yazılım örneği: Sistemde bug yok ama test "bug var!" diyor. Gereksiz yere alarm veriyorsu<, ekip boşuna uğraşıyor. Bir nevi **yanlış alarm.**

- **II. Tip Hata (False Negative) — $\beta$:**
  $H_0$ aslında yanlışken **reddedemedik**, yani farkı yakalayamadık.
  > 🖥️ Yazılım örneği: Sistemde gerçekten bir bug var ama test "her şey normal" diyor. Bug production'a çıkıyor. **Kaçırdığımız bug.**

---

###  Özet

| Hata | ML Karşılığı | Kısa Açıklama |
|------|-------------|---------------|
| **I. Tip ($\alpha$)** | False Positive | Yok olan bir şeyi var sandın |
| **II. Tip ($\beta$)** | False Negative | Var olan bir şeyi kaçırdın |

>   Confusion Matrix'teki **False Positive** ve **False Negative** kavramları tam olarak buradan geliyor. ML'de model değerlendirirken kullanacağımız precision, recall metrikleri aslında bu hata tiplerini ölçüyor. Yani istatistik ve makine öğrenmesi burada birbirine bağlanıyor.

*Teoride H0 hiçbir zaman kabul edilmez.Zaten H0'ı reddetmek için H1'i ortaya sürüyoruz.Eğitim amaçlı sık sık H0'ı doğru kabul edeceğimiz yerler olacaktır.*
