### İş Uygulaması: Anasayfa İçerik Stratejisi Belirlemek

*Bir web sitesinin ortalama ziyaret süresinin uzun olması, hemen çıkış oranı gibi istatistiksel bilgilerin düşük olması o web sitesinin başarı metriklerindendir. Elimizde 3 tarz haber sunma tarzı var (A, B ,C). En iyi sunma tarzının belirlenmesi gerekiyor.*

- **Problem:**
  Anasayfa'da geçirilen süre artırılmak isteniyor

- **Detaylar:**
  - Bir web sitesi için başarı kriterleri: ortalama ziyaret süresi, hemen çıkış oranı vb
  - Uzun zaman geçiren kullanıcıların reklamlara daha fazla tıkladığı ve markaya olan bağlılıklarının arttığı biliniyor.
  - Buna yönelik olarak benzer haberler farklı resimler ya da farklı formatlarda hazırlanarak oluşturulan test gruplarına gösteriliyor.
  - A: Doğal Şekilde, B: Yönlendirici, C: İlgi Çekici


*Bir Veri Bilimci olarak hangi haber sunma tarzının en çok ziyaret edilmeyi sağladığını yapacağımız testlerle belirlememiz gerekiyor. Verinin bize sağlıklı bir şekilde gelmesi ve veriyi doğru bir şekilde yorumlamak gerekir.*

*3 gruptan da veri seti geliyor. Veri seti kullanıcıların A, B, C grup özelliklerini taşıyan web sitesinde ortalama geçirilen süre bilgilerini taşıyor. Veri setleri excel formunda geliyor.*

## Varyans Analizi

$$
\begin{aligned}
H_0&: M_1 = M_2 = M_3 \quad \text{(Grup ortalamaları arasında ist. anl. farklılık yoktur)} \\
H_1&: \text{Fark vardır.}
\end{aligned}
$$


```python
import pandas as pd

A = pd.DataFrame([28,33,30,29,28,29,27,31,30,32,28,33,25,29,27,31,31,30,31,34,30,32,31,34,28,32,31,28,33,29])

B = pd.DataFrame([31,32,30,30,33,32,34,27,36,30,31,30,38,29,30,34,34,31,35,35,33,30,28,29,26,37,31,28,34,33])

C = pd.DataFrame([40,33,38,41,42,43,38,35,39,39,36,34,35,40,38,36,39,36,33,35,38,35,40,40,39,38,38,43,40,42])

dfs = [A, B, C]

ABC = pd.concat(dfs, axis = 1)
ABC.columns = ["GRUP_A","GRUP_B","GRUP_C"]
ABC.head()
```

#### Varsayım Kontrolü

```python
#Normal Varsayımı
from scipy.stats import shapiro
```

```python
shapiro(ABC["GRUP_A"])
```

```python
shapiro(ABC["GRUP_B"])
```

```python
shapiro(ABC["GRUP_C"])
```

*Normal varsayımı: Teorik dağılım ile normal dağılımı arasında anlamlı bir fark yoktur. 3 grup için de p-value değeri alpha değeri olan 0.05'den büyük. Yani H0'ı reddedemiyoruz, ki bu da istediğimiz bir durum.*

```python
#Varyans Homojenliği Varsayımı
from scipy.stats import levene
```

```python
levene(ABC["GRUP_A"], ABC["GRUP_B"], ABC["GRUP_C"])
```

*Varsayım: 3 grup arasında da homojen bir dağılım mevcuttur. H0 bunu söylüyor. p-value 0.05'ten çok büyük. H0 hipotezi reddedilemez.*

*Burada gözlemlerin(A, B, C) birbirinden bağımsız olduğunu kabul ediyoruz.*

#### Hipotez Testi

```python
from scipy.stats import f_oneway
```

```python
f_oneway(ABC["GRUP_A"], ABC["GRUP_B"], ABC["GRUP_C"])
```

```python
#Daha güzel yöntem
print('{:0.5f}'.format(f_oneway(ABC["GRUP_A"], ABC["GRUP_B"], ABC["GRUP_C"])[1]))
```

*p-value değerimiz 0.05'ten çok küçük. H0'ı reddektik. Yani web sitesi tasarımlarının kullanıcılarn web sitesini ziyaret etme süreleri üzerinde kayda değer bir etkisi var.*

```python
ABC.describe().T
```

*ilgi çekici tasarımlara sahip web siteleri kullanıcıların en çok ilgisini çekiyor.*

*Diyelim ki varsayımları sağlayamadık. Ne yapacaktık ? Tabi ki **nonparametrik hipotez testi**.*

```python
from scipy.stats import kruskal
```

```python
kruskal(ABC["GRUP_A"], ABC["GRUP_B"], ABC["GRUP_C"])
```

**(1.70 * e^-12) değeri 0'a çok yakın. Yani H0'ı reddederiz.**
