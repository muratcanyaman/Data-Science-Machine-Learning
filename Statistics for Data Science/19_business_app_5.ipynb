## İş Uygulaması: Ürün Satın Alma Adım Optimizasyonu

- **Problem:**
  Sepete ürün ekleme işlemi sonrasında ödeme ekranında 5 adım vardır ve bu adımların birisi sorgulanmaktadır.

- **Detaylar:**
  - Her adımın 20'şer sn. olması hedefi var. 4. adım sorgulanıyor. Yani 4'üncü adımın da 20 saniye sürdüğünü istatistiksel temelde ispatlamamız gerekiyor.H0 hipotezi bu olacak.
  - Bu durumu test etmek için 100 örnek alınıyor.
  - Örnek standart sapması 5 saniyedir. Örnek ortalaması ise 19 saniyedir.
  

## İş Uygulaması: Ürün Satın Alma Adım Optimizasyonu

**Adım 1:** Hipotezlerin kurulması ve yönlerinin belirlenmesi

$$
H_0: \mu = 20
$$

$$
H_1: \mu \neq 20
$$


*Yukarıdaki hipotezler bize ne diyor?*<br>

*H_0 diyor ki, 4.adımda geçirdiğimiz süre ortalama 20 sn'dir. H_1 de diyor ki sitede zaman geçiren herkesin,yani bilinmeyen ana kitlenin 4.adımda geçirmiş oldukları süre ortalaması 20 sn'den farklıdır.*

**Adım 2:** Anlamlılık düzeyinin ve tablo değerinin belirlenmesi

$$
\alpha = 0{,}05 \quad \quad \frac{\alpha}{2} = 0{,}025
$$

Ztablo tablo olasılık değeri: $0{,}5 - 0{,}025 = 0{,}475$

Ztablo kritik değer $= \pm 1{,}96$(n > 30 olduğu için z tablosu)


*threshold değerini 2'ye bölmemizin sebebi bunun 2 yanlı bir hipotez üzerine çalışmamızdır( < 20 ve > 20). Eğer tek yönlü olsaydı threshold değerini olduğı gibi ilgili tabloda arayacaktık.*

**Adım 3:** Test istatistiğinin belirlenmesi ve test istatistiğinin hesaplanması

$$
z = \frac{\bar{x} - \mu}{\frac{\sigma(s)}{\sqrt{n}}} \quad \quad zhesap = \frac{19 - 20}{5 / \sqrt{100}} = -2{,}00
$$

$n = 100$, standart sapma $= 5$, örnek ortalaması $19$ sn


*Not: Buradaki standart sapma,ana kitleye ait olmayan;Örnekleme ait standart sapma değeridir.Aradaki farka dikkat edelim.*

**Adım 4:** Ztablo ve Zhesap karşılaştırması — $Z_h > Z_t$ ya da $-Z_h < -Z_t$ ise $H_0$ Red

$$
Z_{hesap} = -2{,}00 < Z_{tablo} = -1{,}96 \quad \text{olduğu için } H_0 \text{ reddedilir.}
$$

## 📊 Neden Bu Probleme Hipotez Testi Uyguladık?

Veri biliminde bir iddianın doğru olup olmadığını **sezgiyle değil, veriyle** kanıtlamamız gerekir. Bu problemde elimizde şu durum vardı:

- Hedef: Her adımın **20 saniye** sürmesi
- Gözlem: 100 kullanıcıdan alınan örneklerde 4. adımın ortalaması **19 saniye** çıktı

Burada şöyle bir soru ortaya çıkıyor:

> *"19 ile 20 arasındaki 1 saniyelik fark gerçek bir fark mı, yoksa şans eseri mi oluştu?"*

Eğer sadece ortalamaya bakıp *"19 ≠ 20, demek ki fark var"* deseydik, bu **güvenilir bir çıkarım** olmazdı. Çünkü her örneklem farklı sonuç verir — başka bir 100 kullanıcı alsaydık belki ortalama 20.3 çıkacaktı.

İşte **hipotez testi** tam olarak bunu çözüyor:

- Gözlemlediğimiz farkın **rastlantısal mı yoksa istatistiksel olarak anlamlı mı** olduğunu matematiksel olarak test ediyoruz
- Böylece kararlarımızı **veriye dayalı** ve **ölçülebilir bir güvenilirlik düzeyinde** (%95) alıyoruz

Bu yaklaşım veri biliminde her yerde karşımıza çıkıyor:

| Senaryo | Hipotez Testi Sorusu |
|---------|---------------------|
| A/B Testi | Yeni tasarım dönüşüm oranını gerçekten artırdı mı? |
| Model Karşılaştırma | Yeni modelin accuracy'si eskisinden anlamlı şekilde farklı mı? |
| Süreç Optimizasyonu | Yapılan iyileştirme gerçekten süreyi düşürdü mü? |
| Ürün Analizi | Kullanıcı davranışında gerçek bir değişim var mı? |

Kısacası, hipotez testi bize **"bu fark gerçek mi?"** sorusuna güvenilir bir cevap verir. Veri bilimciler olarak kararlarımızı hislere değil, istatistiksel kanıtlara dayandırmamız gerekiyor.
