## Binom Dağılımı (Binomial Distribution)

Binom dağılımı, bağımsız **n** deneme sonucu **k** başarılı olma olasılığı ile ilgilenildiğinde kullanılan dağılımdır.

### Olasılık Kütle Fonksiyonu (PMF)

$$f(x;\, n, p) = \binom{n}{x}\, p^x\, (1-p)^{n-x}, \qquad x = 0, 1, 2, \dots, n$$

### Beklenen Değer ve Varyans

$$E(X) = np$$

$$Var(X) = np(1 - p)$$

---

**Değişkenler:**

| Sembol | Açıklama |
|--------|----------|
| $n$ | Toplam deneme sayısı |
| $x$ | Başarılı sonuç sayısı |
| $p$ | Bir denemede başarı olasılığı |
| $1 - p$ | Bir denemede başarısızlık olasılığı |
| $\binom{n}{x}$ | Kombinasyon: $n$ denemeden $x$ tanesini seçme sayısı $= \frac{n!}{x!(n-x)!}$ |
| $E(X)$ | Beklenen değer (ortalama) |
| $Var(X)$ | Varyans |

*Binom dağılımı Bernoulli Dağılımı'ın n deneme sonucu ortaya çıkan dağılımıdır diyebiliriz.*

## Binom Dağılımı - Örnek Soru

**Soru:** Bir madeni para 4 kere atılıyor. 2 kere yazı gelmesi olasılığı nedir?

### Formül

$$f(x;\, n, p) = \binom{n}{x}\, p^x\, (1-p)^{n-x}, \qquad x = 0, 1, 2, \dots, n$$

### Çözüm

- $n = 4$ (para 4 kere atılıyor)
- $x = 2$ (2 kere yazı gelmesini istiyoruz)
- $p = 0.50$ (yazı gelme olasılığı)

$$f(2;\, 4,\, 0.50) = \binom{4}{2}\, 0.50^2\, (1 - 0.50)^{4-2} = 0.375$$

**Sonuç:** 4 atışta 2 kere yazı gelme olasılığı **%37.5**'tir.
