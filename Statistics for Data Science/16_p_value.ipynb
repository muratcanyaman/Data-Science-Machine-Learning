## p-value Nedir? Ne Anlama Gelir?

Hipotez testlerini öğrenirken kafamı en çok karıştıran kavramlardan biri **p-value** oldu. Ama aslında mantığı çok basit. Şöyle düşünüyorum:

> **p-value:** $H_0$ doğruyken, elimizdeki veri kadar **uç (extreme)** bir sonuç elde etme olasılığıdır.

Yani p-value bana diyor ki:
> *"Eğer $H_0$ gerçekten doğruysa, senin gördüğün bu sonucu sadece şans eseri görme ihtimalin şu kadar."*

- **p-value küçükse** → Bu sonucu şans eseri görmem çok zor demek → $H_0$ muhtemelen yanlış → **$H_0$'ı reddederiz** ✅
- **p-value büyükse** → Bu sonucu şans eseri görmem gayet mümkün → $H_0$'ı reddetmek için yeterli kanıt yok → **$H_0$'ı reddedemeyiz** ❌

---

## Neden $p < 0.05$ Olduğunda Seviniriz?

$0.05$ değeri istatistikte yaygın olarak kullanılan **anlamlılık düzeyi ($\alpha$)**'dir. Bu aslında bir **eşik değer (threshold)**.

$$
p < \alpha \quad \Rightarrow \quad H_0 \text{ reddedilir (istatistiksel olarak anlamlı)}
$$

$$
p \geq \alpha \quad \Rightarrow \quad H_0 \text{ reddedilemez}
$$

Genelde $\alpha = 0.05$ alınır, yani:

| p-value | Karar | Anlamı |
|---------|-------|--------|
| $p < 0.05$ | $H_0$ **reddedilir** ✅ | Gözlemlediğimiz fark gerçek, şans eseri değil |
| $p \geq 0.05$ | $H_0$ **reddedilemez** ❌ | Yeterli kanıt yok, fark şanstan da kaynaklanabilir |

Seviniriz çünkü $p < 0.05$ demek → **"Bulduğumuz sonuç istatistiksel olarak anlamlı, bu fark gerçek!"** demek.

> 🖥️ Yazılım dünyasından örnek:
> Bir A/B testi yaptım. Yeni butonun tıklanma oranı eskisinden %3 daha yüksek çıktı.
> - $p = 0.02$ → Bu artış gerçek, şans değil! Yeni butonu deploy edebilirim. 
> - $p = 0.35$ → Bu artış şanstan kaynaklanıyor olabilir. Daha fazla veri toplamalıyım. 

---

## Dağılım Testlerinde Neden $H_0$'ın Reddedilmemesini İsteriz?

Normalde hipotez testlerinde $H_0$'ı reddetmek isteriz ama **dağılım testlerinde (normallik testi gibi)** tam tersi geçerli. Sebebi şu:

Normallik testlerinde (Shapiro-Wilk, Kolmogorov-Smirnov gibi):

$$
H_0: \text{Veri normal dağılıma uyar}
$$

$$
H_1: \text{Veri normal dağılıma uymaz}
$$

Yani $H_0$'ın kendisi **"veri normal dağılıyor"** diyor. Eğer biz verinin normal dağıldığını **varsaymak** istiyorsak (ki birçok istatistiksel test ve ML algoritması bunu gerektirir), o zaman $H_0$'ın **reddedilmemesini** isteriz.

| Sonuç | Anlamı | Biz ne isteriz? |
|-------|--------|----------------|
| $p < 0.05$ → $H_0$ reddedildi | Veri normal dağılmıyor | ❌ İstemeyiz |
| $p \geq 0.05$ → $H_0$ reddedilemedi | Veri normal dağılıyor olabilir | ✅ Bunu isteriz |

> 💡 Kısacası, hipotez testlerinde her zaman **$H_0$'ın ne dediğine** bakmamız gerekiyor.
> Eğer $H_0$ bizim işimize yarayan şeyi söylüyorsa → reddedilmemesini isteriz.
> Eğer $H_0$ bizim karşıt görüşümüzü söylüyorsa → reddedilmesini isteriz.

---

## 🔑 Özet

- **p-value**, $H_0$ doğruyken bu sonucu şans eseri görme olasılığı.
- **p < 0.05** → Sonuç anlamlı, $H_0$ reddedilir.
- **Dağılım testlerinde** $H_0$ = "veri normal" olduğu için, $H_0$'ın reddedilmemesi bizim lehimize.
- Her zaman önce **"$H_0$ ne diyor?"** diye sormalıyız, sonra karar vermeliyiz.
