## Hipotez Testi Adımları

Bir hipotez testi yaparken izlememiz gereken 5 temel adım var:

---

**Adım 1:** Hipotezlerin kurulması ve yönlerinin belirlenmesi

> İlk olarak $H_0$ ve $H_1$ hipotezlerimizi yazıyoruz. Testimizin yönünü de belirliyoruz: iki yönlü mü, sağ kuyruk mu, sol kuyruk mu?

**Adım 2:** Anlamlılık düzeyinin ve tablo değerinin belirlenmesi

> Anlamlılık düzeyimizi ($\alpha$) belirliyoruz. Genelde $\alpha = 0.05$ alırız. Bu düzeye karşılık gelen **kritik tablo değerini ($Z_t$)** istatistik tablolarından buluruz.

**Adım 3:** Test istatistiğinin belirlenmesi ve test istatistiğinin hesaplanması

> Hangi testi kullanacağımıza karar veriyoruz (Z testi, T testi vb.) ve elimizdeki veri ile **test istatistiğini ($Z_h$)** hesaplıyoruz.

**Adım 4:** Hesaplanan test istatistiği ile alfa'ya karşılık gelen tablo değerinin karşılaştırılması

> Hesapladığımız test istatistiğini ($Z_h$) tablo değeri ($Z_t$) ile kıyaslıyoruz:
>
> $$Z_h > Z_t \implies H_0 \text{ reddedilir}$$
>
> $$Z_h \leq Z_t \implies H_0 \text{ reddedilemez}$$

**Adım 5:** Yorum

> Son olarak sonuçlarımızı yorumluyoruz: $H_0$ reddedildiyse ne anlama geliyor, reddedilemediyse ne anlama geliyor? Bulgumuzu iş problemi bağlamında açıklıyoruz.

---

> 💡 Pratikte Python ile çalışırken genelde **Adım 2 ve 4'ü manuel yapmamıza gerek kalmıyor**. `scipy.stats` kütüphanesi bize direkt **p-value** döndürüyor, biz de sadece $p < \alpha$ mı diye kontrol ediyoruz. Ama bu adımları bilmek, arka planda ne olduğunu anlamamızı sağlıyor.
