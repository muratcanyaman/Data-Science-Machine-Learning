### Varyans Analizi

*İki ya da daha fazla grup ortalaması arasında istatistiksel olarak anlamlı farklılık olup olmadığı öğrenilmek istenildiğinde kullanılır.*

#### Hipotezler

$$
\begin{aligned}
H_0&: \mu_1 = \mu_2 = \mu_3 \\
H_1&: \text{Eşit değillerdir (en az birisi farklıdır)}
\end{aligned}
$$


#### Varyans Analizi: Test İstatistiği

$$
F_s = \frac{MS_{(between)}}{MS_{(within)}}
$$

| Sembol | Açıklama |
|--------|----------|
| $F_s$ | Hesaplanan F test istatistiği |
| $MS_{(between)}$ | Gruplar arası ortalama kareler toplamı (gruplar arasındaki varyans) |
| $MS_{(within)}$ | Gruplar içi ortalama kareler toplamı (grupların kendi içindeki varyans) |

#### Varyans Analizi: Varsayımlar

- Gözlemlerin birbirinden bağımsız olması (grupların)
- Normal dağılım
- Varyans homojenliği
