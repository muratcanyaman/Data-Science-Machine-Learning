### İki Örneklem Oran Testi

*İki oran arasında karşılaştırma yapmak için kullanılır.*

#### Hipotezler

| İki Yönlü Test | Sağ Kuyruk Testi | Sol Kuyruk Testi |
|----------------|------------------|------------------|
| $H_0: P_1 = P_2$ | $H_0: P_1 \leq P_2$ | $H_0: P_1 \geq P_2$ |
| $H_1: P_1 \neq P_2$ | $H_1: P_1 > P_2$ | $H_1: P_1 < P_2$ |


#### Test İstatistiği
$$
Z_h = \frac{(p_1 - p_2)}{\sqrt{p(1-p)\left(\frac{1}{n_1} + \frac{1}{n_2}\right)}}
$$

| Sembol | Açıklama |
|--------|----------|
| $p_1$ | 1. grubun gözlenen oranı |
| $p_2$ | 2. grubun gözlenen oranı |
| $p$ | Havuzlanmış (pooled) oran: $p = \frac{n_1 p_1 + n_2 p_2}{n_1 + n_2}$ |
| $n_1$ | 1. grubun örneklem büyüklüğü |
| $n_2$ | 2. grubun örneklem büyüklüğü |
| $Z_h$ | Hesaplanan test istatistiği |



#### Varsayımlar

- $n_1 > 30$
- $n_2 > 30$
