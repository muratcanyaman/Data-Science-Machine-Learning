## Poisson Dağılımı (Poisson Distribution)

Belirli bir zaman aralığında belirli bir alanda nadiren rastlanan olayların olasılıklarını hesaplamak için kullanılır.

### Olasılık Kütle Fonksiyonu (PMF)

$$f(x, \lambda) = \frac{\lambda^x \, e^{-\lambda}}{x!}, \qquad x = 0, 1, 2, \dots, n$$

### Beklenen Değer ve Varyans

$$E(X) = \lambda \qquad Var(X) = \lambda$$

---

**Değişkenler ve Sabitler:**

| Sembol | Açıklama |
|--------|----------|
| $x$ | İlgilenilen olay sayısı (kaç kez gerçekleştiği) |
| $\lambda$ (lambda) | Belirli bir zaman aralığında olayın ortalama gerçekleşme sayısı |
| $e$ | Euler sayısı (doğal logaritma tabanı, $e \approx 2.71828$) |
| $x!$ | $x$ faktöriyel ($x! = x \times (x-1) \times \dots \times 1$) |
| $E(X)$ | Beklenen değer (ortalama) — Poisson'da $\lambda$'ya eşittir |
| $Var(X)$ | Varyans — Poisson'da yine $\lambda$'ya eşittir |

> **Not:** Poisson dağılımında beklenen değer ve varyans birbirine eşittir ve her ikisi de $\lambda$ değerine eşittir. Bu, Poisson dağılımının ayırt edici özelliklerinden biridir.

*Gözlem sayısının çok yüksek ve beklenen sonucun gelme olasılığının çok küçük olduğu zamanlarda kullanılır.*

*Nadir olaylardan kastımız nedir?*

*İki şart vardır: Rassal deneme sayısının(n) 50'den büyük olması ve n * p değerinin(lambda) 5'ten küçük olması.Bu iki şartı sağlayan olaylar nadir olay kabul edilir.*

*Uyarı !*

*- Rassal denemeler iki sonuçlu olmalı.*<br> 
*- Aynı koşullar altında gerçekleşmelidir.*<br>
*- Rassal denemeler birbirinden bağımsız olmalıdır.*

## Poisson Dağılımı Örnek

Bir üniversitede 5000 not girişinde 5 tane notun yanlış girilmesi olasılığı nedir?

Dağılımın Poisson olduğu biliniyor ve $\lambda = 0.2$

### Çözüm

- $x = 5$ (yanlış girilmesi beklenen not sayısı)
- $\lambda = 0.2$ (ortalama yanlış not girişi oranı)

$$f(5;\, 0.2) = \frac{0.2^5 \, e^{-0.2}}{5!} = 0.00000218328201$$
