### Değişken Dönüşümü

*Aslında her standardizasyon işlemi de bir değişken dönüşümü işlemi. Fakat asıl sormamız gereken soru dönüşüm sonrasında değişken yapısının bozulup bozulmadığı sorusu. Standardizasyon işlemlerinde genelde yapısal değişiklik olmaz. Değişken dönüşümlerinde bu ihtimal daha yüksektir.*

```python
import seaborn as sns
df = sns.load_dataset("tips")
df.head()
```

#### 0 - 1 Dönüşümü

```python
from sklearn.preprocessing import LabelEncoder
lbe = LabelEncoder()
```

```python
df["new_sex"] = lbe.fit_transform(df["sex"])
```

```python
df.head()
```

#### "1 ve Diğerleri(0) Dönüşümü"

```python
df.head()
```

```python
import numpy as np
df["new_day"] = np.where(df["day"].str.contains("Sun"), 1, 0)
```

```python
df.head(100)
```

#### Çok Dikkat !

*Çok sınıflı (nominal) bir kategorik değişkene Label Encoding uygulandığında, sınıflar arasında gerçekte var olmayan yapay bir sıralama ve mesafe ilişkisi oluşturulmuş olur. Örneğin, yukarıdaki gibi bir Label Encoding uygulandığında Thur = 0, Fri = 1, Sat = 2, Sun = 3 şeklinde bir kodlama yapılır. Bu durumda model, Thur ile Sun arasındaki farkın (|0 − 3| = 3), Thur ile Fri arasındaki farktan (|0 − 1| = 1) üç kat daha büyük olduğunu varsayar. Oysa günler arasında böyle bir büyüklük veya uzaklık ilişkisi yoktur. Sınıf sayısı arttıkça bu yapay mesafe farkları da büyür ve özellikle mesafe tabanlı algoritmalarda (KNN, K-Means, SVM vb.) ciddi performans kayıplarına yol açar. Bu nedenle, doğal bir sıralaması olmayan kategorik değişkenlerde Label Encoding yerine **One-Hot Encoding**, **Target Encoding** veya **Frequency Encoding** gibi alternatif yöntemler tercih edilmelidir.*
