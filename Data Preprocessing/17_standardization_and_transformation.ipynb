## Veri Standartlaştırma & Değişken Dönüşümü

### Standartlaştırma

```python
import numpy as np
import pandas as pd

V1 = np.array([1,3,6,5,7])
V2 = np.array([7,7,5,8,12])
V3 = np.array([6,12,5,6,14])

df = pd.DataFrame(
    {"V1" : V1,
     "V2" : V2,
     "V3" : V3}
)

df = df.astype(float)
df
```

```python
from sklearn import preprocessing
```

```python
preprocessing.scale(df)
```

### Normalizasyon

```python
preprocessing.normalize(df)
```

### Min-Max Dönüşümü

```python
scaler = preprocessing.MinMaxScaler(feature_range = (10,20))
```

```python
scaler.fit_transform(df)
```

### Binarize Dönüşüm

```python
#Değişken Dönüştürme işlemidir ve bilginin kendisini yapısal olarak etkiler.
```

```python
binarazer = preprocessing.Binarizer(threshold = 5).fit(df)
```

```python
binarazer.transform(df)
```

### 0-1 Dönüşümü

```python
import seaborn as sns
tips = sns.load_dataset("tips")
df = tips.copy()
df_l = df.copy()
```

```python
df_l.head()
```

```python
df_l["yeni_sex"] = df["sex"].cat.codes
```

```python
df_l.head()
```

```python
lbe = preprocessing.LabelEncoder()
```

```python
df_l["daha_yeni_sex"] = lbe.fit_transform(df_l["sex"])
```

```python
df_l.head()
```

### "1 ve Diğerleri"(0) Dönüşümü

```python
#İkiden fazla sınıf bulunduran değişken dönüşümlerinde kullanılır.
```

```python
df.head()
```

```python
df_l.head()
```

```python
df_l["yeni_gun"] = np.where(df_l["day"].str.contains("Sun"), 1, 0)
```

```python
df_l.head(20)
```

### Çok Sınıflı Dönüşüm 

```python
lbe = preprocessing.LabelEncoder()
```

```python
df_l["daha_yeni_gun"] = lbe.fit_transform(df_l["day"])
```

```python
df_l
```

### One-Hot Dönüşümü ve Dummy Değişken Tuzağı

### Sürekli Değişkeni Kategorik Değişkene Çevirmek

```python
est = preprocessing.KBinsDiscretizer(n_bins = [3,2,2], encode = "ordinal", strategy = "quantile").fit(dff)
```

```python
est.transform(dff)[0 : 10]
```
