#### Eksik Veri Analizi

*İncelenen veri setindeki gözlemlerde eksiklik bulunmasını ifade eder. Veri girişleri, tablo birleştirme veya değişken dönüştürme işlemleri gibi işlemler sırasında ortaya çıkabilmektedir.*

*Eksik değerler yukarıdaki tabloda olduğu gibi **NA** , **NaN** ya da **Null** şeklinde temsil edilir.Eksik değerleri ortadan kaldırmanın birkaç yolu var: Ya eksik değere sahip gözlemlerin direkt olarak veri setinden çıkarılması ya da ortalama değerin verilmesi gibi. Her türlü eksik verileri ortadan kaldırmanın yıkıcı sonuçlar doğuracağının bilincinde olmalıyız.*

*Eksik değere sahip gözlemlerin veri setinden direkt çıkarılması ve rassallığının incelenmemesi yapılacak istatistiksel çıkarımların, modelleme çalışmalarının güvenilirliğini düşürecektir.
(Alpar, 2011)*

*Eksik gözlemlerin veri setinden direkt çıkarılabilmesi için veri setindeki eksikliğin bazı durumlarda kısmen bazı durumlarda tamamen rastlantısal olarak oluşmuş olması gerekmektedir. Eğer eksiklikler değişkenler ile ilişkili olarak ortaya çıkan yapısal problemler ile meydana gelmiş ise bu durumda yapılacak silme işlemleri ciddi yanlılıklara sebep olacaktır.
(Tabachnick ve Fidell,1996)*

#### Eksik Veriyi Direkt Silmenin Zararları

*1.**Veri setindeki eksikliğin yapısal olup olmadığının bilinmesi gerekir!***

*Yukarıdaki gözlemi inceleyelim. Bir müşterinin kredi kartına sahip olmadığı bilgisi var. Kredi kartı harcaması yapmaması durumu NA olarak girilmiş. Yani kredi kartına sahip olmayan müşterilerin harcamaları NA olarak girilmiş.Çünkü bu müşterilerin kredi kartları yok ve harcama yapılmamış yani herhangi bir veri aktarımı vs. gibi bir işlem sonucu hatalı giriş durumu yok. Bu NA değerleri ortalama harcama değeri ile doldurulursa bu durum gürültü oluşumuna, bundan dolayı da yanlılığa sebep olacak. Bir değişkenin başka bir değişkene bağımlı olmasına bağlı eksiklik olması söz konusu.*

*2.**Na her zaman eksiklik anlamına gelmez!***

*Yukarıda kredi kartına sahip bir müşterinin kredi kartı harcaması durumu NA olarak belirtilmiş. Burada müşterinin kredi kartı bilgisi var. Yani müşterinin kart bilgileri işlenmiş fakat müşteri hiç harcama yapmamış. Yani yapılan harcama NA olarak temsil ediliyor. NA burada eksiklik anlamına gelmiyor. Sadece yapılan harcama olan  0; NA ile temsil ediliyor. Burada yapısal bir eksiklik yok. Sadece mevcut durum farklı bir şekilde temsil edilmiş.*

*3.**Bilgi Kaybı***

*Yukarıda n müşteriden ve 100 değişkenden oluşan bir veri setimiz var.Müşteri1 gözlemine ait 100'üncü değişkende NA değeri var, diğer 99'u dolu. Yani müşteri ile ilgili 100 özellikten 99'u biliniyor. Bu gözlemi silersek 99 tane farklı özelliği de silmiş oluyoruz. Yani veri setimizde bilgi kaybı meydana gelir.*

### Eksik Veri Türleri Nelerdir?

***Tümüyle Rastlantısal Kayıp:** Diğer değişkenlerden ya da yapısal bir problemden kaynaklanmayan tamamen rastgele oluşan gözlemler. En çok karşılaşmak istediğimiz eksik veri türüdür.*<br>
***Rastlantısal Kayıp:** Diğer değişkenlere bağlı olarak oluşabilen veri türüdür.*<br>
***Rastlantısal Olmayan Kayıp:** Göz ardı edilemeyecek olan ve yapısal problemler ile ortaya çıkan eksiklik türü. En sevilmeyen kayıp türüdür.*
