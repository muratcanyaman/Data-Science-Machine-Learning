## Aykırı Gözlem Nedir?

*Veride genel eğilimin oldukça dışına çıkan ya da diğer gözlemlerden oldukça farklı olan gözlemlere **Aykırı Gözlem** denir.*

### Aykırı Değer(Outlier) Nedir?

*Aykırılığı ifade eden nümerik değere **Aykırı Değer** denir.*

*Buna göre aykırı değeri barındıran gözlem birimine **Aykırı Gözlem** diyebiliriz.*

*Genellenebilirlik kaygısı ile oluşturulan kural setlerini ya da fonksiyonları yanıltır. Bu da **yanlılılığa** sebep olur.*

*Yukarıdaki resimde bir veri setindeki iki değişken arasındaki ilişki grafik ile gösteriliyor. Aykırı gözlemsiz grafikte x ve y değişkeni arasındaki ilişkiyi temsil eden mavi eğri negatif eğimlidir yani bu iki değişken arasındaki ilişki normalde bu şekildedir. Fakat aykırı gözlem içeren grafikte x ve y değişkenleri arasındaki ilişkiyi gösteren eğri pozitif hale gelmiş. 2-3 tane aykırı gözlem yüzünden verinin genel yapısını temsil eden eğrinin yönü değişmiş. İşte buna **Yanlılık** diyoruz.*

### Neye Göre Kime Göre Aykırı Gözlem ? 

*Aykırı gözlemi verinin genel eğilimin dışına çıkan gözlem olarak tanımlamıştık. Peki veri setinin genel eğiliminin dışına çıkmayı nasıl tanımlarız ?*

#### 1. Sektör Bilgisi

*Örneğin bir ev fiyat tahmin modelinde 1000 metrekarelik evleri modellemeye almamak. Türkiye'de evlerin %99'u 1+1, 2+1, 3+1 evlerden oluşuyor. 1000 metrekare evler sadece %1'lik kesimi oluşturuyor. Eğiteceğimiz Makine Öğrenmesi Algoritmaları genellenebilirlik kaygısı taşıyor. Yani biz bu %1'lik kesimi de genele dahil etmemiz yanlılığa sebep olacak yani model yanlış tahminlerde bulunacak.*

*NOT: Eğer kurulan modelin bir genelleme kaygısı varsa; zaten çok seyrek olan senaryolar ve genele uymayan yapılar çalışmanın dışında bırakılmalıdır.*

#### 2. Standart Sapma Yaklaşımı

*Bir değişkenin ortalamasının üzerine aynı değişkenin standart sapması hesaplanarak eklenir. 1,2 ya da 3 standart sapma değeri ortalama üzerine eklenerek ortaya çıkan bu değer eşik değer olarak düşünülür ve bu değerden yukarıda ya da aşağıda olan değerler aykırı değer olarak tanımlanır.*

*Örneğin araba fiyatlarını tutan değişken ortalaması 100.000, standart sapması 20.000 olsun. Ortalama üzerine 2 standart sapma değeri daha ekleyip(140.000) bu değeri eşik değer olarak kabul edebiliriz. Bu değerin altında kalan değerler de aykırı değer olur.*

#### 3.Z-Skoru Yaklaşımı

*Standart Sapma yöntemine benzer. Değişken standart normal dağılıma uyarlanır, yani standartlaştırılır. Sonrasında -örneğin- dağılımın sağından ve solundan -+2.5 değerine göre eşik değer konulur ve bu değerin üzerinde ya da altında olan değerler aykırı değer olarak işaretlenir.*

#### 4.Boxplot(interquartile range - IQR) Yöntemi 

*En sık kullanılan yöntemlerden birisidir. Değişkenin değeri küçükten büyüğe sıralanır. Çeyrekliklerine (yüzdelikleine) yani Q1, Q3 değerlerine karşılık değerler üzerinden bir eşik değer hesaplanır ve bu eşik değere göre aykırı değer tanımı yapılır.*
