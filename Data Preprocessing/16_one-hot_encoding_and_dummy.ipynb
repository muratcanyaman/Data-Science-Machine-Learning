## One-Hot Dönüşümü ve Dummy Değişken Tuzağı

*Bir önceki bölümde kategorik değişkene label encoding yaptıktan sonra ortaya çıkan ve gerçekte var olmayan yapay bir sıralama ve mesafe ilişkisinden bahsetmiştik. Bunun çözümü One-Hot Dönüşümü'dür. Bu dönüşümü yaparken de ortaya Dummy değişken tuzağı dediğimiz bir problem ortaya çıkıyor. Bunun çözümünden de bahsedeceğiz.*

*Veri setimizi ve üzerinde yaptığımız işlemleri hatırlayalım:*

```python
import seaborn as sns
df = sns.load_dataset("tips")
df.head()
```

```python
#0-1 Dönüşümü
```

```python
from sklearn.preprocessing import LabelEncoder
lbe = LabelEncoder()
```

```python
df["new_sex"] = lbe.fit_transform(df["sex"])
```

```python
from sklearn.preprocessing import LabelEncoder
lbe = LabelEncoder()
```

```python
import pandas as pd
df_one_hot = pd.get_dummies(df, columns = ["sex"], prefix = ["sex"], dtype = int)
```

```python
df_one_hot.head()
```

*Orijinal veri setindeki "sex" kategorik değişken sınıflarının her birini nümerik değişkenlere dönüştürdük ve bu değişken sınıflarını 1 ve 0 olarak tanımladık. Örneğin indeks 1'deki gözlem biriminde müşteri erkek, indeks 0'da müşteri kadın.*

*Burada dummy değişken tuzağı ortaya çıkıyor. Bir veri setindeki değişkene one-hot dönüşümü uyguladıktan sonra ortaya çıkan değişkenler birbirleri üzerinden oluşturulabiliyorsa bu dummy değişken tuzağı oluyor. Yani burada sex_Male değişkenine bakarak sex_Female değişkenini oluşturabiliriz ya da tam tersini uygulayarak oluşturabiliriz. Yani bu iki değişken şu an aynı bilgiyi taşıyor. Bu işlem sonrasında dönüşüm uyguladığımız kategorik değişkenin sınıf sayısından 1 eksik dummy değişken olması gerekir. Değişkenin orijinal halinde olduğu gibi iki değil bir sınıf olmalı.*

#### Peki Bu İşlemi Yapmamızdaki Temel Sebep Ne ? 

*Burada dönüşüm uyguladığımız kategorik değişken sınıflarının etkisini veri setinde daha hissedilebilir olmasını sağladık. Algoritmaya ilgili değişkenin ağırlığını kazandırma imkanı sağlıyor. Örnek üzerinden açıklayalım : 10 sınıflı bir kategorik değişken var ve bu değişken veri setimizin kilit değişkeni olsun ve algoritma bu değişken üzerinden yazılsın. Bu 10 sınıflı değişkene dönüşüm uyguladığımızda yeni veri setimizde bu 10 sınıf da nümerik olarak temsil edilecek. Bu işlem ilgili kategorik değişkenin gözlem bazında veri setindeki ağırlığını arttırıyor. Aynı zamanda nümerik dönüştürme işlemini de aradan çıkartmış oluyoruz. Yani aslında algoritmaya diyoruz ki bak bu değişkenin şu kadar sınıfı var ve bu değişken benim veri setimin kilit noktası. Senin için nümerik hale de getirdim kendini buna göre eğit demiş oluyoruz.*

*One-Hot Dönüşümü'nde üzerinde dönüşüm yapmak istediğimiz kategorik değişkenin sınıf sayısı arttıkça bu sınıfların birbirleri üzerinden dönüşüm yapılabilirliği de güçleşiyor. Yani çok sınıflı kategorik değişkenlerin dönüşümü yerinde uygulunacak bir hamledir. Fakat "sex" kategorik değişken örneğinde olduğu gibi dummy tuzağı olup olmama durumu her zaman kontrol edilmelidir.*
