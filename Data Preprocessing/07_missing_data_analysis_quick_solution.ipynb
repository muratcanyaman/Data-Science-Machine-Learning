#### Eksik Veri Analizi

```python
import numpy as np
import pandas as pd
V1 = np.array([1,3,6,np.nan,7,1,np.nan,9,15])
V2 = np.array([7,np.nan,5,8,12,np.nan,np.nan,2,3])
V3 = np.array([np.nan,12,5,6,14,7,np.nan,2,31])
df = pd.DataFrame(
    {"V1":V1,
     "V2":V2,
     "V3":V3})
df
```

#### Eksik Değerlere Erişmek

```python
#boş değerlerin gösterilmesi
df.isnull().sum()
```

```python
#boş olmayan değerlere erişim
df.notnull().sum()
```

```python
#toplam boş değer bilgisine erişim
df.isnull().sum().sum()
```

```python
df.isnull()
```

```python
# en az bir değeri NaN olan bütün gözlem birimlerinin getirilmesi
df[df.isnull().any(axis = 1)]
```

```python
# bütün değerleri dolu olan gözlem birimlerinin getirilmesi
df[df.notnull().all(axis = 1)]
```

```python
#hiçbir şey bilmesek bile alternatif gösterim
df[df["V1"].notnull() & df["V2"].notnull() & df["V3"].notnull()]
```

#### Eksik Değerlerin Silinmesi

```python
df.dropna(inplace = False)
#kalıcı olarak silme
#df.dropna(inplace = True)
```

```python
df
```

#### Basit Değer Atama

```python
df["V1"]
```

```python
df["V1"].mean()
```

```python
df["V1"].fillna(df["V1"].mean())
```

```python
df["V2"].fillna(0)
```

*Onlarca değişkene sahip bir veri setinde böyle çalışmak anlamsız. Bu işlemleri otomatize etmemiz gerek. Burada fonksiyonel programlama devreye giriyor.*

```python
#boş değerlere otomatik atama işlemi
```

```python
df.apply(lambda x: x.fillna(x.mean()), axis = 0)
```

*apply() fonksiyonunun değişkenler üzerinde çalıştığını hatırlayalım. x, burada değişkenleri temsil ediyor. axis = 0 parametresi de her bir değişken için satır satır dolaşmamız gerektiğini söylüyor.*
