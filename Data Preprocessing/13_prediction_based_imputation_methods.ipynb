### Tahmine Dayalı Atama Yöntemleri

*Veri setindeki NaN değerleri Makine Öğrenmesi Algoritmaları kullanarak tahmine dayalı bir şekilde doldurabiliriz.Eksik değere sahip değişken, diğer değişkenlerce modellenip doldurulma işlemi gerçekleştirilir.*

```python
import seaborn as sns
import missingno as msno
df = sns.load_dataset("titanic")
df = df.select_dtypes(include = ["float64", "int64"])
print(df.head())
df.isnull().sum()
```

```python
#KNN algoritmasına göre değer atama
```

```python
import numpy as np
np.infty = np.inf  # NumPy 2.0 uyumluluk yaması

from ycimpute.imputer import knnimput
```

*knnimput fonksiyonu parametre olarak numpy array'i bekler. O yüzden dataframe nesnesini numpy array'ine çevirmemiz gerek. Fakat ondan önce de değişken adlarını tutmalıyız.*

```python
var_names = list(df)
```

```python
n_df = np.array(df)
```

```python
n_df[0:10]
```

```python
n_df.shape
```

```python
dff = knnimput.KNN(k = 4).complete(n_df)
```

```python
type(dff)
```

```python
type(dff)
```

```python
dff.isnull().sum()
```

```python
#RandomForests kullanarak değer atama
```

```python
import seaborn as sns
import missingno as msno
df = sns.load_dataset("titanic")
df = df.select_dtypes(include = ["float64", "int64"])
```

```python
df.isnull().sum()
```

```python
var_names = list(df)
```

```python
n_df = np.array(df)
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.ensemble import RandomForestRegressor

imputer = IterativeImputer(
    estimator=RandomForestRegressor(n_estimators=100, random_state=42),
    random_state=42
)

dff = imputer.fit_transform(n_df)
```

```python
type(dff)
```

```python
dff = pd.DataFrame(dff, columns = var_names)
```

```python
type(dff)
```

```python
dff.isnull().sum()
```
