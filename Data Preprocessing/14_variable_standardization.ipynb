### Değişken Standardizasyonu (Veri Standardizasyonu)

*Değişken Standardizasyonu Nedir ? Değişken Standardizasyonu değişkenin kendi içindeki bilgi yapısının bozulmadan belirli bir standarta oturtulmasıdır. Değeri değişir,ama temsil ettiği yayılım ve dağılım bilgisi değişmez. Örneğin 10 değerinin veri setinde bir değişken içerisinde küçükten büyüğe doğru sıralandığında 80.sırada olduğunu varsayalım. Bu 10 değerinin bulunduğu değişken; standardizasyon işlemine tabi tutulduktan sonra değeri -1.5786859 gibi bir değer olacak. Fakat sıralama işlemi sonunda bu değer yine 80.sırada olacaktır. Değişkenin ortalama,medyan, standart sapma gibi değerleri değişecektir fakat bu değerin değişken içerisindeki yayılım ve dağılım bilgisi değişmeyecektir.*

```python
import numpy as np
import pandas as pd
V1 = np.array([1,3,5,6,7])
V2 = np.array([7,7,5,8,12])
V3 = np.array([6,12,5,6,14])
df = pd.DataFrame(
    {"V1" : V1,
     "V2" : V2,
     "V3" : V3}
)

df = df.astype(float)
df
```

#### Standardizasyon

```python
from sklearn import preprocessing
```

```python
preprocessing.scale(df)
```

```python
df.mean()
```

#### Normalizasyon

```python
preprocessing.normalize(df)
```

#### Min - MaX Dönüşümü

```python
scaler = preprocessing.MinMaxScaler(feature_range = (10,20))
```

```python
scaler.fit_transform(df)
```
