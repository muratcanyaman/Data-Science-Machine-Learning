### Kategorik Değişken Kırılımında Değer Atama 

*Kategorik Değişken Kırılımında Değer Atama ne demek? Bir şirkette çalışan bütün çalışanların bilgilerinin tutulduğu bir veri seti düşünelim. Bu veri setinde kişinin maaş ve çalıştıkları departman bilgileri olsun. Üst düzey bir departmanda yüksek maaşla çalışan bir çalışanın maaş bilgisine atama yapacağız. Eğer şirketin genelini kapsayan bir ortalama değeri atarsak bu maaş bilgisi yanıltıcı olacaktır.Çünkü bu çalışanın aldığı maaş genel ortalamanın çok yukarısında olabilir.O yüzden bu gibi değer farkının yüksek olduğu durumlarda değişken kırılımında değer ataması yaparız. Yani veri setini indirgeyerek bu veri seti üzerinde çalışırız.*

```python
import numpy as np
import pandas as pd
V1 = np.array([1,3,6,np.nan,7,1,np.nan,9,15])
V2 = np.array([7,np.nan,5,8,12,np.nan,np.nan,2,3])
V3 = np.array([np.nan,12,5,6,14,7,np.nan,2,31])
V4 = np.array(["IT","IT","IK","IK","IK","IK","IK","IT","IT"])
df = pd.DataFrame(
    {"maas" : V1,
     "V2" : V2,
     "V3" : V3,
     "departman" : V4}
)
df
```

```python
df.groupby("departman")["maas"].mean()
```

```python
df["maas"].fillna(df.groupby("departman")["maas"].transform("mean"))
```

*Burada **transform()** fonksiyonu kritik öneme sahip.Gruplama sonucu elde edilen çıktının boyutunu orijinal dataframe ile aynı tutuyor.agg() veya mean() gibi indirgeme yapmıyor.Böylece **maas** değişkenine her bir departmanın maaş ortalamasını broadcast edebiliyoruz. Amaç çıktı ile dataframe uyumsuzluğunu önlemek.*
