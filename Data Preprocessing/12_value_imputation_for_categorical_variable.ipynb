### Kategorik Değişkenler için Eksik Değer Atama

```python
import numpy as np
import pandas as pd
V1 = np.array([1,3,6,np.nan,7,1,np.nan,9,15])
V4 = np.array(["IT",np.nan,"IK","IK","IK","IK","IK","IT","IT"], dtype = object)


df = pd.DataFrame(
    {"maas" : V1,
     "departman" : V4}
)

df
```

*Kategorik değişkenlerde en mantıklı değer atama yöntemi en çok tekrar eden değerin atanmasıdır(mod).*

```python
df["departman"].mode()
```

```python
df = df["departman"].fillna(df["departman"].mode()[0])
```
