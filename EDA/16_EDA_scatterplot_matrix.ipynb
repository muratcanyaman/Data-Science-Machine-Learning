### Scatterplot Matrisi

*Sayısal değişkenler arasındaki ilişkiyi bir matris ile ifade etmek amacıyla Scatterplot Matrisleri kullanılır.*

```python
import seaborn as sns
iris = sns.load_dataset("iris")
df = iris.copy()
df.head()
```

```python
df.dtypes
```

```python
df.shape
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue = "species")
```

```python
sns.pairplot(df, hue = "species", markers = ["o", "s", "D"])
```

```python
sns.pairplot(df, kind = "reg")
```

```python
sns.pairplot(df, kind = "reg", hue = "species",)
```
