### Heatmap

*Elimizdeki verileri daha yapısal, daha geniş perspektif ile incelemek istediğimiz zaman Heatmap bu konuda bize çok yardımcı oluyor.  Özellikle Zaman Serisi verilerinin incelenmesinde oldukça kullanışlıdır.*

### Veri Seti Hikayesi

```python
import seaborn as sns
flights = sns.load_dataset("flights")
df = flights.copy()
df.head()
```

```python
df.shape
```

```python
df["passengers"].describe()
```

*Bazı durumlarda heatmap() metodu dataframe'i olduğu gibi haritalayamayabilir. Bunun için dataframe üzerinde bazı yapısal değişiklikler yapmak gerekebilir. Örneğin yukarıdaki veri setinde ayları satır, yılları da sutün olarak düzenlemek gibi bir düzenleme yapmak gerekir.*

```python
df = df.pivot(index = "month", columns = "year", values = "passengers")
```

```python
df
```

```python
sns.heatmap(df)
```

*Daha detaylı bir heatmap için:*

```python
sns.heatmap(df, annot = True, fmt = "d", linewidth = 0.5)
```
