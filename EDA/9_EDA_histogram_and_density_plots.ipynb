## Histogram ve Yoğunluk

*Çoğunlukla sayısal değişkenlerin dağılımını ifade etmek için kullanılan görselleştirme teknikleridir.*

```python
import seaborn as sns
diamonds = sns.load_dataset("diamonds")
df = diamonds.copy()
df.head()
```

```python
sns.displot(df.price, kde = False)
```

```python
df["price"].describe()
```

```python
sns.displot(df.price, bins = 10, kde = False)
```

*Olasılık-Yoğunluk dağılımını göstermek için aşağıdaki gibi bir kod yazabiliriz:*

```python
sns.kdeplot(df.price)
```

```python
sns.kdeplot(df.price, fill = True)
```
