## Lojistik Regresyon (Teori)

*Amaç sınıflandırma problemi için bağımlı ve bağımsız değişkenler arasındaki ilişkiyi tanımlayan doğrusal bir model kurmaktır.*

*Genellikle x'in her bir kategoriye ait olma olasılığı ile ilgilenilir.*

*Sınıf tahminlerinin performansı; yanlış sınıflandırma oranı ya da doğru sınıflandırma oranı ile ölçülür.*

*Peki bu sınıflandırma işini neden Lojistik Regresyon ile yapıyoruz ? Yanıt değişkenimizin iki değer aldığı durumlarda (0 ve 1) Doğrusal Regresyon Modeli kuramaz mıyız ?*

*Evet, aslında Doğrusal Regresyon bağımlı değişkenimiz iki sınıflı ise (0 ve 1 gibi) kullanılabilir. Fakat ikiden fazla değer aldığı durumlarda kullanılamaz. İki değer aldığı durumlarda ise aldığımız bu değerler bazen -1,2,3 gibi değerler de olabilir. Bu durum da çözülebilir. 1'den fazla değerler 1; 0'dan küçük değerler 0 olarak kabul edilebilir. Fakat bunun da getirdiği bazı problemler var.*

*Lojistik Regresyonda kullanılan fonksiyon ile tahmin değerlerinin 0 ile 1 arasında olması durumu garanti altına alınmış olunur. İlgilendiğimiz sınıfın 1 olması olasılığı ile ilgili olasılık değerleri hesaplanmış olur.*

## Özellikler

- *Bağımlı değişken kategoriktir.*
- *Adını bağımlı değişkene uygulanana **Logit Dönüşümü**'nden alır.*
- *Doğrusal Regresyon'da aranan varsayımlar burada aranmadığı için daha esnek bir kullanılabilirliği vardır.*
- *Bağımlı değişkenin 1 olarak tanımlanan değerinin gerçekleşme olasılığı hesaplanır. Yani modele yollayacağımız gözlem biriminin hangi sınıfa ait olabileceği olasılığıyla ilgileniyoruz. Dolayısıyla bağımlı değişken ile ilgilenmiyoruz.*
- *Lojistik fonksiyonu sayesinde üretilen değerler 0-1 arasında olur.*
