## Destek Vektör Sınıflandırıcısı (Teori)

*Amaç iki sınıf arasındaki ayrımın optimum olmasını sağlayacak hiper düzlemi bulmaktır.*

*Hiper düzlem 2 ya da 3 boyutlu düzlemi ifade ediyor. Marjin ise hiper düzleme en yakın veri noktalarını ifade eden alandır.*

*Optimum hiper düzlem nasıl bulunur ? İki doğru arasında nokta olmayacak şekilde aralarındaki mesafenin maksimum tutulmaya çalışılmasıyla bulunur.*

*Yukarıdaki görselde iki boyutlu bir marjin görüyoruz. Görselde ik sınıf kusursuz bir şekilde ayrılabilmiş görünüyor. Gerçek hayatta veriler çok boyutlu olarak gelmekte ve görseldeki gibi doğrusal bir ayrımı kolay bir şekilde yapabilmek oldukça zordur.*

*Bu durumda **Doğrusal Olmayan SVM** dediğimiz durum ortaya çıkmakta ve **kernel-trick** dediğimiz yöntemler ile bu ayrıştırma işlemi gerçekleştirilmektedir.*

<div align="center">
    
</div>

*Yukarıdaki görselin sol tarafındaki grafikte kırmızı ve mavi sınıflar, üzerinden doğrusal bir çizgi çekilmesine rağmen ayrılamamış. Yani veri yapısı doğrusal bir şekilde sınıflara ayrılmaya müsait değil.*

*Buna çözüm olarak ikinci görseldeki gibi boyut ekleme yöntemi uygulanmış. Mavi sınıfa ait verilerin yukarıya çekildiğini düşünün. Kırmızı sınıfa ait veriler ise aşağıda kalıyor. Araya bir düzlem çekilerek çok boyutlu verimiz sınıflarına ayrılmış oluyor. Tabi bu ayrıştırma işlemleri birtakım fonksiyonlar ile gerçekleştiriliyor (radyal-bases).*
