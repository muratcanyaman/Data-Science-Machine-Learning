### KNN Basamakları

- *Komşu sayısını belirle (K)*
- *Bilinmeyen nokta ile diğer tüm noktalar ile arasındaki uzaklıkları hesapla*
- *Uzaklıkları sırala ve belirlenen k sayısına göre en yakın olan k gözlemi seç*
