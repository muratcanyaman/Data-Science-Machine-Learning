## Light GBM (Teori)

*Light GBM, XGBoost'un eğitim süresi performansını arttırmaya yönelik geliştirilen bir diğer GBM türüdür.*

- *Daha performanslı*
- *Level-wise büyüme stratejisi(XGBoost) yerine Leaf-wise büyüme stratejisi.*
- *Breadth-first Search (BFS) yerine Depth-first Search (DFS)*

*Veri setindeki değişken ve gözlem sayılarının çok olduğu senaryolarda C++ dilinde yazılmış olsa da XGBoost'un eğitimi çok uzun sürmekte. Tam bu noktada LightGBM ortaya çıkmıştır. Daha büyük veri setlerinde daha performanslı, eğitim süresi daha düşük. Temelinde yine karar ağaçları var.*

*XGBoost ile farklılaştığı nokta şu: XGBoost değişkenleri ayırmak ve farklı bölmeleri keşfetmek için karar ağaçlarını kullanıyor. LightGBM ise split'lere odaklanıyor. Ağaç yapılarındaki bölme işlemleri düşünüldüğünde XGBoost geniş kapsamlı bir ilk arama yaparken LightGBM derinlemesine bir arama yapıyor. Bu anlamda LightGBM performans olarak daha ön plana çıkıyor.*
