## Apache Hadoop

*Apache Hadoop açık kaynak kodlu, güvenilir, ölçeklenebilir paralel hesaplama yazılımı projesidir.*

*Büyük veri problemini yazılımsal olarak çözen bir framework. Birden fazla bilgisayarın eş zamanlı olarak çalışmasını sağlayan bir yazılım mimarisidir.*

- *Büyük veri teknolojisinin temelini oluşturur.*
- *Geleneksel yöntemler ile etkin olarak işlenmesi mümkün olmayan verilerin işlenebilmesine olanak sağlamaktadır.*
- *Bir bilgisayar kümesinin belirli bir işi yapmak için tek bir bilgisayar gibi hareket etmesini sağlamaktadır.*

### Apache Hadoop Bileşenleri

- *Hadoop Common : Diğer Hadoop modüllerini destekleyen ortak gereksinimlerdir.*
- *Hadoop Dağıtık Dosya Sistemi (HDFS) : FAT32, NTFS gibi bir dosya sistemi. Çok büyük verileri dağıtık bir şekilde tutmamızı sağlar.*
- *Hadoop YARN : Kaynak yönetimi ve iş planlaması için kullanılır.*
- *Hadoop MapReduce : En önemli bileşen. Aynı ağdaki dağıtık bilgisayar kümeleri üzerinde büyük veri analizi işlemleri yapılabilmesi için geliştirilmiş bir programlama modelidir.*

## MapReduce Nedir ? 

*Bilgisayar topluluğunun belirli bir amaç için birlikte hareket edebilmesini sağlayan  programlama modeli / yazılımıdır.*

<div align="center"></div>

*Yukarıdaki görseldeki bir iş problemi üzerinden MapReduce'nin ne olduğunu daha detaylı bir şekilde inceleyelim. Elimizde bir metin var. Bu metin içerisinde yüz milyonlarca hayvan ismi geçiyor (dog, cat, duck gibi). Bu hayvanların her birinden kaç kere bahsediliyor bunun sayısını hesaplamak istiyoruz.*

*Mapping ve Reducing iki ana aşamadır. Mapping aşamasında HDFS üzerindeki girdi verileri işlenir. Metnimiz map fonksiyonu ile satır satır işlenir ve yeni veri parçacıkları oluşturulur. Metnimiz yukarıdaki 3 farklı makineye split edilmiş. Satır satır işlenen bu metinde hayvan sayıları işaretlenir. Shuffling basamağında her bir hayvan aynı türde olacak şekilde tekrar sayılıyor, düzenleniyor. Reducing adımında bu hayvan sayıları toplanıyor, yani tekleniyor. Merged basamağında ise bu metinde her bir hayvan adı kaç defa geçmiş bunun çıktısını alıyoruz.*

## Apache Hadoop Çözüm Olarak Yeterli mi ? 

- *Disk tabanlı çalışan bir modeldir.(Veri ambarı olarak çok avantajlı)*
- *Her MapReduce görevinde diskten okuma ve diske yazma işlemi yapılır.*
- *İteratif işlemler zaman alır ve kaynakları meşgul eder. Veri Madencili'ni, Makine Öğrenmesi algoritmalarının çalışmasını pek sevmez.*

<div align="center"></div>

*Yukarıda Mapreduce Disk Bazlı Çalışma Modeli'ni görüyoruz. Bir Makine Öğrenmesi problemi ile ilgilendiğimizi varsayalım. Ağaç yapılarından boosting metodolojisi ile çalıştığımızı düşünelim. İlk iterasyonda hata hesaplamalarını yaptık. Bu hesapları dikkate alıp ikinci bir model geliştirdik. Sonra tekrar hata hesaplamaları yaptık, tekrar yeni bir model geliştirdik. Her iterasyonda bu hesaplar yapılırken diske yazma- diskten okuma işlemleri gerçekleştiriliyor. Çok büyük veriler ile çalışırken okuma-yazma işlemleri çok uzun sürer. Bu da zaman ve performans kaybıdır.*

*Tam da bu probleme çözüm olarak **Apache Spark** geliştirilmiştir. Hadoop'a alternatif değil, Hadoop üzerine kurulmuştur.
