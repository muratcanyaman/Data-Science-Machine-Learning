## Büyük Veri Nedir ? 

*Kısaca tanım yapmak gerekirse geleneksel yöntemlerle işlenemeyen verilere büyük veri denir.*

*Büyük Veri aslında bir problemdir. Geleneksel yöntemler ile işlenememesi, boyut ve tür çeşitliliğinin artması ile veriyi tutma, saklama, işleme aşamalarında zorluklar çıkarması ile  büyük veri bir problemdir.*

### Büyük veri bir problem ise çözümü nedir ? 

*Bu problemin çözümü birden fazla bilgisayarın bir araya gelerek tek bir bilgisayar gibi hareket etmesidir. Yani bir işi yapmak için birden fazla bilgisayarın tek bir bilgisayar gibi davranmasıdır. Bu davranma hem donanımsal hem de yazılımsaldır.*

### Bu probmelin çözülmesi bize ne kazandırdı ? 

*Veri Analitiği alanında yeni ufuklar açtı, hesaplama gücünün artmasıyla makine öğrenmesi algoritmalarının performans gücü arttı. Daha büyük miktar ve çeşitteki verilerin kullanılması ile veriden faydalı bilgi çıkarma süreci için çok önemli bir kaynak / araç sağlanmış oldu.*

### Büyük veri bir problemdi ve çözüldü

*Bugün elektrik ve suyun varlığını tartışmadığımız gibi gelecekte verinin büyüklüğünü ve bu veriyi işlemenin zorluklarını da tartışmıyor olacağız.*

*Büyük veri büyük bir nimettir ! Büyük veri, veri odaklı çalışanlar için çok büyük bir zenginlik, yeni keşifler için zemin ve dünyaya veri penceresinden olan bakış açımızı değiştiren bir vizyondur.*

## Büyük Verinin Bileşenleri

<div align="center"></div>

- *Ortaya çıkan veri işleme güçlüğü verinin hacmi, çeşitliliği ve hızı ile alakalıdır.*
- *Büyük veriyi ifade eden özellikler hacim, çeşitlilik ve hızdır.*
- *Çok yaygın bir yanılgı: Bu üç özelliğin (3V) birlikte olduğunda ancak bu tür verilere büyük veri denileceği yanılgısıdır. Bir veriyi ilişkisel veri tabanında tutamıyorsak o veri büyük veridir diyebiliriz.*
- *Büyük veri araçları veriden faydalı bilgi çıkarma süreçleri için çok güçlü bir araçtır(Hadoop, Spark...).*
