## Apache Spark

*Apache Spark küme üzerinde hızlı ve genel amaçlı bilgi işleme sistemidir. Apache Hadoop üzerine inşa edilmiştir. MapReduce veri işleme mimarisinin bir alternatifidir.*

- *Mapredcue modelinde yer alan disk bazlı çalışma sisteminin yarattığı maliyetlerden dolayı ortaya çıkmıştır.*
- *Apache Hadoop'a göre 100 kat daha hızlı çalışmaktadır.*
- *Java, Scala, Python ve R ile uygulama geliştirilebilir.*
- *Genelleştiricidir : Spark SQL, Spark MLlib, Spark Streaming, GraphX aynı uygulamada kullanılabilir.*

### Apache Spark Bileşenleri

- *Spark Core ve RDD's : Hafıza yönetimi, görev dağıtımı, hata kurtarma, dosya işlemlerine erişim.*
- *Spark SQL : Spark programındaki yapılandırılmış verinin sorgulanmasını sağlar.*
- *Spark MLlib : Ölçeklenebilir Makine Öğrenmesi Kütüphanesi.*
- *Spark Streaming : Akan verinin ölçeklenebilir, yüksek hacimli ve hata toleranslı olarak işlenmesini sağlayan bir API uzantısıdır.*
- *GraphX : Paralel tabanlı hesaplamalar için kullanılan bir kütüphane*

## Dayanıklı Dağıtık Dataset'ler (RDD's)

<div align="center"></div>

*Apache Spark'ın paralel veri işleme mimarisidir. Hadoop'taki Mapreduce mimarisinin dezavantajlarını ortadan kaldırmak için geliştirildi.*

*Veri bellek içerisinde tutularak (In-Memory) paralel bir şekilde işlenir.*

*3 temel aşamadan oluşur : **RDDs**, **Transformation** ve **Action***

- *Önce RDD's oluşturulur. Bu RDDs oluşturma işlemi herhangi bir görev olabilir, Veri oluşturma vs. RDDs üretilen işlemi anlar ve askıda bekletir. Diske yazma işlemi gerçekleştirmez. Bu işlemi RAM üzerinde gerçekleştirir.*
- *Transformation aşamasında bir RDDs üzerinden başka bir RDDs üretilebilir. Örneğin belirli bir değişkenin belirli bir değerden büyük olması durumuna göre veriyi filtreledik. Bu RDDs'e bir transformation işlemi gerçekleştiğini gösterir. Filtreleme işleminden sonra değişken seçimi işlemi yaptık. Bu da bir Transformation işlemidir. Yani veri üzerinde çeşitli işlemler yapılırken diske yazılmaz. Üzerinde işlem yapılan veri bellek içerisinde tutulur.*
- *Action aşamasında Transformation aşamasında üretilen RDDs'ler işlenir ve toplanarak sonuçlar elde edilir.*

*Veri üzerindeki operasyonlar bitene kadar diske yazılmaz. Buna **Lazy Evaluation** deniyor.*

*Transformation -> Hadoop'taki Mapping*<br>
*Action -> Hadoop'taki Reducing*
