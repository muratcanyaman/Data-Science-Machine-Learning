## Hiyerarşik Kümeleme (Teori)

*Amaç gözlemleri birbirlerine olan benzerliklerine göre alt kümelere ayırmaktır.*

<div align="center"></div>

*Yukarıdaki görsel ülke eyaletlerinin hiyerarşik kümeleme analizi sonucu elde edilen çıktısını göstermekte. Eyaletler 4 ana alt kümeye ayrılmış. 4 kümenin her biri yine kendi içerisinde alt kümelere ayrılmış. Yukarıdan aşağıya doğru bir kümeleme işlemi söz konusu ise **Divisive Bölümleyici** yaklaşımı kullanılır. Eğer gözlemlerin benzlerliklerine göre bir kümeleme işlemi yapılıyorsa **Agglomerative Birleştirici** yöntemi uygulanır. Yani gözlem benzlerliklerine bakılarak en alttan en yukarıya bir kümeleme söz konusu.*

*Hiyerarşik Kümeleme nasıl uygulanır ?*

**Birleştirici Kümeleme(Agglomerative):**

*Başlangıçta gözlem sayısı kadar küme vardır.*<br>

- *Adım 1: Veri setinde birbirine en yakın 2 gözlem bulunur.*
- *Adım 2: Bu iki nokta bir araya getirilerek yeni bir gözlem oluşturulur. Yani artık veri seti ilk birleşimdeki gözlemlerden oluşmaktadır.*
- *Adım 3: Aynı işlem tekrarlanarak yukarı doğru çıkılır. Yani iki kümenin birleşiminden oluşan bu yeni kümeler aynı şekilde birbirlerine benzerliklerine göre tekrar birleştirilir. Bu işlem tüm gözlemler tek bir kümede toplanana kadar bu işlemler tekrar edilir.*

*Birbirlerine yakın noktalar uzaklık ölçümleri kullanılarak belirlenir. Öklit uzaklığı, Manhattan uzaklığı, korelasyon vs.*

**Bölümleyici Kümeleme(Divisive):**

*Başlangıçta tek bir küme vardır, o da veri setinin kendisidir.*

- *Adım 1: Tüm gözlemlerin bir arada olduğu küme iki alt kümeye ayrılır*
- *Adım 2: Oluşan yeni kümeler birbirlerine benzemeyen alt kümelere bölünür.*
- *Adım 3: Aynı işlem gözlem sayısı kadar küme elde edilinceye kadar devam edilir.*

### Hiyerarşik vs Hiyerarşik Olmayan vs Karar Ağaçları

- *Hiyerarşik yöntemlerde küme sayısına Dendogram sonuçlarına bakılarak karar verilirken, hiyerarşik olmayan yöntemlerde küme sayısı uygulama yapılmadan önce belirlenir.*
- *Hiyerarşik kümeleme yöntemlerinde veri seti gözlemler ya da değişkenler bazında kümeleme işlemine sokulabilirken hiyerarşik olmayan kümeleme yöntemlerde sadece gözlemlerin kümelenmesi mümkündür.*
- *Karar Ağaçlarından farkı: karar ağaçlarında ayırma işlemi hedef değişkene göre yapılırken burada bağımlı değişken olmadığı için gözlemler bağımsız değişkenler üzerinden yapılan uzaklık hesaplarına göre kümelere ayrılır.*
