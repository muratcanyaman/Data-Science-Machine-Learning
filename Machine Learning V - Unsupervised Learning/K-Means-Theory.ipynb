## K-Means

*Gözetimsiz Öğrenme: Elimizde bağımlı değişken olmadığı zaman gözlem birimlerinin nasıl anlamlı hale getirilebilir çabasıdır. Veri setinde bağımlı değişkenin ne olduğunu bilmiyoruz ama bağımsız değişkenlerin birbirleri arasındaki ilişkiyi ortaya çıkarmaya çalışıyoruz. K-Means de bu yöntemlerden bir tanesi.*

*K-Means'te amaç, gözlemleri birbirlerine olan benzerliklerine göre kümelere ayırmaktır.*

*Benzerlik matrislerini kullanarak, çeşitli uzaklık metrikleri ile uzaklık hesapları yaparak gözlemler veya değişkenler kümelenmeye çalışılmaktadır. Oluşturulan bu kümelerin kendi içerisinde homojen, birbirlerine göre heterojen olması amaçlanır.*

*Hiyerarşik ve Hiyerarşik Olmayan Kümeleme Yöntemleri şeklinde iki başlıkta incelenir. K-Means Hiyerarşik Olmayan Kümeleme Yöntemi'dir.*

*K-Means'te nihai amaç: Oluşturulan kümelerin birbirleri içerisinde benzerliğinin yüksek, birbirlerine göre benzerliğinin düşük olmasını sağlamaya çalışmaktır.*

## K-Means Adımları

- *Adım 1: Küme sayısı belirlenir.*
- *Adım 2: Rastgele K merkez seçilir.*
- *Adım 3: Her gözlem için k merkezlere uzaklıklar hesaplanır ve gözlemler kendisine en yakın k merkezine atanır.*
- *Adım 4: Her gözlem en yakın olduğu merkeze yani kümeye atanır.*
- *Adım 5: Atama işlemlerinden sonra oluşan kümeler için tekrar merkez hesaplamaları yapılır.*
- *Adım 6: Bu işlem belirlenen bir iterasyon adedince tekrar edilir ve küme içi hata kareler toplamlarının toplamının (total within-cluster variation) minimum olduğu gözlemlerin kümelenme yapısı nihai kümelenme olarak seçilir.*

*Kümeleme işlemi tamamlanıncaya kadar gözlemler farklı farklı kümelere girip çıkabilmektedir. İyi bir K-Means çıktısı elde edebilmek için, yani kümeler arası benzerliğin düşük kümeler içi benzerliğin yüksek olduğu bir çıktı elde edene kadar K-Means adımları iterasyon şeklinde devam eder.*
