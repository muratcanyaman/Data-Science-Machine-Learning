## Temel Bileşen Analizi (Teori)

*Temel fikir, çok değişkenli verinin ana özelliklerini daha az sayıda değişken / bileşen ile temsil etmektir. Diğer bir ifade ile: küçük miktarda bir bilgi kaybını göze alıp değişken boyutunu azaltmaktır.*

<div align="center"></div>

*İndirgenen bileşenler arasında korelasyon yoktur.*

*Değişken gruplarının varyanslarını ifade eden özdeğerler ile veri setindeki değişkenler gruplandırılır, gruplar arasında en fazla varyansa sahip gruplar en önemli gruplardır ki bunlar **Asal Bileşenler**'dir.*
