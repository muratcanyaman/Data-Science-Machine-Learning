### Özellikler

- *Aşırı öğrenmeye karşı dirençlidir.*<br>
- *Yanlıdır fakat varyansı düşüktür.(Bazen yanlı modelleri daha çok tercih ederiz.)*<br>
- *Çok fazla parametre olduğunda EKK'ye göre göre daha iyidir.*<br>
- *Çok boyutluluk lanetine karşı çözüm sunar.*<br>
- *Çoklu doğrusal problemi olduğunda etkilidir.*<br>
- *Tüm değişkenler ile model kurar. İlgisiz değişkenleri modelden çıkarmaz, katsayıları sıfıra yaklaştırır.*<br>
- *λ kritik roldedir. Formüldeki iki terimin de görece etkilerini kontrol etmeyi sağlar.*<br>
- *λ için iyi bir değer bulunması önemlidir. Bunun için CV yöntemi kullanılır.*<br>
