#### statsmodels ile model kurulumu

 ## Model Tuning / Model Doğrulama

*Çoklu Doğrusal Regresyon'da dışsal olarak görülebilecek bir parametre olmadığı için burada Model Tuning'i aslında Model Doğrulama olarak ele alacağız.*

```python
df.head()
```

```python
X = df.drop('sales', axis = 1)
y = df['sales']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 144)

lm = LinearRegression()
model = lm.fit(X, y)
```

```python
# tahmin edilen eğitim hatası
```

```python
np.sqrt(mean_squared_error(y_train, model.predict(X_train)))
```

```python
# tahmin edilen test hatası
```

```python
np.sqrt(mean_squared_error(y_test, model.predict(X_test)))
```

```python
#doğrulanmış eğitim hatası
```

```python
np.sqrt(-cross_val_score(model,
                X_train,
                y_train, 
                cv = 10,
                scoring = "neg_mean_squared_error")).mean()
```

```python
#doğrulanmış test hatası
```

```python
np.sqrt(-cross_val_score(model,
                X_test,
                y_test, 
                cv = 10,
                scoring = "neg_mean_squared_error")).mean()
```

*Burada dikkat etmemiz gereken birkaç husus var: Birincisi eğitim hatasının test hatasından farklılaşıyor olacağı ve eğitim hatasının test hatasının kötü bir tahmincisi olduğudur.*

*cross validation işlemi ile eğitim hatasını ve test hatasını doğrulamaya çalışıyoruz yani valide edilmiş hata değerleri daha güvenilirdir.*

*Bundan sonra model oluştururken eğitim setini valide ederek hatalara erişmek ve valide edilmiş modelin katsayılarını yani parametrelerini kullanarak yeni modeller üretmek olacak.*
