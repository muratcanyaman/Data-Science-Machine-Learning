## Yapay Sinir Ağları (Teori)

*İnsan beyninin bilgi işleme şeklini referans alan sınıflandırma ve regresyon problemleri için kullanılabilen kuvvetli makine öğrenmesi algoritmalarından bir tanesidir.*

<div align="center">
    
</div>

*Girdiler aslında bizim bağımsız değişkenlerimizin değeridir. Her bir değişken değerimiz bir ağırlık ile çarpılıyor. Bu ne anlama geliyor ? Yaptığımız bu işlem değişkenlerin değerlerinin çıktıya olan etkisinin kontrol edilmesi demek. Yani regresyonda olduğu gibi bağımsız değişkenlerin bağımlı değişken üzerindeki etkisini belirliyoruz.*

*Amacımız her zaman hataları minimum yapacak matematiksel bağıntıyı bulmak. Ağırlıklar en optimum değerlerine ulaşana kadar geriye yayılımlı bir şekilde sürekli güncelleniyor. Bu güncelleme işlemi gerçek değerler ile tahmin edilen değerlerin karşılaştırılıp hataların dağıtılması ile yapılıyor. Ağırlıklar güncellenirken model öğrenme işlemi gerçekleştiriyor. Mantık aynı.*

*Yukarıdaki görsel bir yapay sinir hücresini temsil ediyor. Bunların bir araya gelmesiyle de **Yapay Sinir Ağları** oluşuyor.*

<div align="center">
    
</div>

*Yukarıda 3 katmanlı bir yapay sinir ağı örneği var. Gerçek değerler ile tahmini değerler arasındaki fark hesaplanıp hata oranı olabildiğince azaltılmaya çalışılana dek ağırlıkların güncellendiğini ve bu işlemin geriye yayılımlı bir şekilde olduğunu söylemiştik. Bu işlemler gizli katmanda gerçekleşiyor. Gizli katman veri setinin boyutuna ve problemin tipine göre çok daha fazla sayıda olabilir. Çok daha karmaşık yapay sinir ağları mevcut. Resimde sadece basic bir ağ gösteriliyor.*

<div align="center">
    
</div>

$h_k(x)$ (Gizli Katman): Girdi verilerini ağırlıklandırarak (weight) özellik çıkarımı yapar.<br>
$g(u)$ (Sigmoid): Aktivasyon fonksiyonudur; veriyi 0-1 arasına sıkıştırarak modele doğrusal olmayan (non-linear) bir yapı katar.<br>
$f(x)$ (Çıktı): Gizli katmandan gelen bilgileri birleştirerek nihai tahmini üretir. Yani yapay sinir ağının genel denklemi oluyor.<br>
Loss Function (Alt Formül): Modelin hata payını (MSE) hesaplar. Sondaki $\lambda$ terimleri ise L2 Regülarizasyonudur; ağırlıkları sınırlayarak(cezalandırma) modelin ezberlemesini (overfitting) engeller.
