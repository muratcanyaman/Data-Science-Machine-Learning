## Optimizasyon Problemi

Grafik, Modelin tahmin doğrusunun etrafında ±ε genişliğinde bir "tolerans bandı"'nı temsil ediyor.
Band içinde kalan noktalar sıfır hata sayılır. Dışarıda kalan noktalar için
ξ (xi) ve ξ* ile gösterilen kayıp miktarları tanımlanır.<br>

Minimizasyon problemi'nde modelin hem basit (düzgün) kalması hem de bant dışı
noktalara verilen toplam cezanın küçük olması amaçlanır. C parametresi bu iki
hedef arasındaki dengeyi kontrol eder.<br>

Kısıtlar ya da kısi dediğimiz şey her noktanın tahmin bandından ne kadar saptığı, ilgili kayıp
değişkeni (ξ veya ξ*) kadar olabilir. Kayıp değişkenleri negatif olamaz.

*Yani SVR bize der ki, sana öyle bir optimum tahmin fonksiyonu üreteceğim ki bu fonksiyonun tahmin ettiği değerler ile gerçek değerler arasındaki fark belirli bir epsilon ve kısi değerinden daha fazla uzaklıkta olamayacak.*
