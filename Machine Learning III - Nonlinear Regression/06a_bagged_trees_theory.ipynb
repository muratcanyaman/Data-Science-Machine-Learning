## Bagged Trees (Teori)

*Temeli bootstrap yöntemi ile oluşturulan birden fazla karar ağacının ürettiği tahminlerin bir araya getirilerek değerlendirilmesine dayanır.*

*Topluluk öğrenme yöntemini benimseyen bir modeldir.*

*T adet ağaç oluşturulur, her bir ağaç için n adet gözlem birimi seçilir. Her bir ağaç için bir tahmin değeri hesaplanır ve değerlendirilir.*

*Bu T adet ağacın birbirleriyle bağıımlı olma durumu söz konusu değildir.*

*örneklerin her bir ağaç oluşturulmasında kullanılması şansı eşittir.* 

*Aşırı öğrenmeye karşı dayanıklıdır.*
