## Karar Ağaçları (CART) - Teori

*Amaç veri seti içerisindeki karmaşık yapıları basit karar yapılarına dönüştürmektir. Heterojen veri setleri belirlenmiş bir hedef değişkene göre homojen alt gruplara ayrılır.*

*Yukarıda hitters veri setine ait bir karar ağacı görseli var. Veri setini belirli karar kurallarına göre heterojen yapıdan homojen yapıya dönüştürmeye çalışıyoruz. Bu kurallar entropi, chi-kare değeri gibi değerlere göre belirlenir. Deneyim Yılı ve Atış Sayısı değişkenlerine göre maaş hesaplama işlemi yapacağımızı düşünelim. Deneyim Yılı 4.5'ten küçükse maaşı ortalama 5.11 olur diyoruz. Eğer Deneyim Yılı 4.5'tan büyükse veya eşitse atış sayısına bakacağız. Atış Sayısı 117.5'ten küçükse maaşı ortalama 6.00 olacak. Eğer Atış Sayısı 117.5'ten büyükse veya eşitse maaşı 6.74 olacak diyoruz.*

*Deneyim yılı ve Atış sayısı değişkenleri ağacın iç düğüm noktalarıdır. Diğer 3 düğüm de terminal düğüm ya da yaprak adını alır. Yaprak değerleri aslında ortalama regresyon değerleridir. Mesela Deneyim yılı 4.5'ten az ve fazla diye 2 gruba ayırdık. 4.5'ten az olan grubun ortalama değeri 5.11 oluyor.*

*Bir ağaç yapısında değişkenleri iç düğüm olarak yerleştirirken o bağımsız değişkenin bağımlı değişkeni açıklayıcılık ve ayırt edicilik anlamında ne kadar iyi olduğunu göz önünde bulundururuz. Yukarıdaki ağaçta bağımlı değişken olan maaş değişkenini; Deneyim Yılı Değişkeni Atış Sayısı değişkeninden açıklayıcılık ve ayırt edicilik bakımından daha iyidir. Bu yüzden daha tepededir. Bağımsız değişkenlerin önem derecesi yukarıdan aşağıya azalır.*

*Yukarıdaki görselde başka bir karar ağacı modelinin eksen üzerinde gösterimi verilmiş. Soldaki if - else öbeği aslında bizim kural setimiz. A tahmincisi 1.7'den büyük veya eşit ise diğer if öbeğine giriyoruz. Hem tahminci A 1.7'den büyük hem de tahminci B 202.1'den büyük veya eşit ise çıktı olarak 1.3 değeri geliyor. Sadece Tahminci A değeri sağlanıyorsa 5.6 değerini elde ediyoruz. İki tahminci değeri de sağlanmıyorsa çıktımız 2.5 oluyor.*

*Yalnız bu değerlerin birden çok gözlem değerinin ortalaması olduğunu aklımızdan çıkarmayalım. Eksenin tamamını heterojen, 3 parçaya ayrılmış bölgelerin her birini homojen alt parçalar olarak düşünebiliriz.*

*Çok fazla dallanmanın veri setini çok iyi bir şekilde temsil edildiğini söyleyebiliriz. Dışarıdan bir gözlem birimi geldiğinde train seti içerisinde çok başarılı bir şekilde tahmin yapabilir. Fakat başka bir veri seti ile karşılaştığı zaman bu aşırı dallanmadan dolayı modelimiz tahmin yapmada başarısız olabilir. Yani çok fazla dallanma ile modeli overfit etmiş oluyoruz.*

*Bunun çözümü *budama* yöntemidir. Bir karmaşıklık parametresi belirlenerek ağacın belirli bir yerden itibaren dallanması durduruluyor. Karmaşıklık parametresi Cross Validation ile belirlenir. Ağacın dallanmasının nerede durdurulması gerektiğini de hata kareler toplamının düşüş durumuna göre belirli bir eşik değerin altına geldiğimizde budama işlemini sonlandırma gerçekleştirilir.*

*Genellenebilirlik kaygımız yok ise CART çok iyi bir tercih olacaktır.*
