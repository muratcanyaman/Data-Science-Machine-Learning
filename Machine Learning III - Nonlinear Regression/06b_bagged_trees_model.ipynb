## Bagged Trees (Model)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import BaggingRegressor
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
hit = pd.read_csv("hitters.csv")
df = hit.copy()
df = df.dropna()
dms = pd.get_dummies(df[["League", "Division", "NewLeague"]])
dms = dms.astype(int)
y = df["Salary"]
X_ = df.drop(["Salary", "League", "Division", "NewLeague"], axis = 1).astype("float64")
X = pd.concat([X_, dms[["League_N", "Division_W", "NewLeague_N"]]], axis = 1)
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y,
                                                    test_size = 0.25,
                                                    random_state = 42)
```

```python
bag_model = BaggingRegressor(bootstrap_features = True)
bag_model.fit(X_train, y_train)
```

## Tahmin

```python
y_pred = bag_model.predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_test, y_pred))
```

*Yukarıdaki değer birnevi bütün ağaçların hata değerinin ortalaması. Bütün ağaçları tek tek kontrol edip senin tahminin nedir diye bir sorgu yapma şansımız da var. Kötü tahminci olan ağaçları tek tek inceleme şansımız var. Yani mevcut ağaçların herhangi birinin yaptığı hata değerini gözlemleyip kötü tahmin yapan ağaçlaro optimizasyon işlemleri uygulayabiliriz.*

```python
sec_y_pred = bag_model.estimators_[1].fit(X_train, y_train).predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_test, sec_y_pred))
```

*Baya beceriksiz bir ağaca denk geldik !*

## Model Tuning

```python
bag_model = BaggingRegressor(bootstrap_features = True)
bag_model.fit(X_train, y_train)
```

```python
bag_params = {"n_estimators" : range(2, 20)}
```

```python
bag_cv_model = GridSearchCV(bag_model, bag_params, cv = 10)
```

```python
bag_cv_model.fit(X_train, y_train)
```

```python
bag_cv_model.best_params_
```

```python
#final model
```

```python
tuned_bag_model = BaggingRegressor(n_estimators = 19, random_state = 45)
```

```python
tuned_bag_model.fit(X_train, y_train)
```

```python
#final test hatasi
```

```python
y_pred = tuned_bag_model.predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_test, y_pred))
```
