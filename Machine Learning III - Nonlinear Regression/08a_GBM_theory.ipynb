## Gradient Boosting Machines (Teori)

*Adaboost'un sınıflandırma ve regresyon problemlerine kolayca uyarlanabilen genelleştirilmiş versiyonudur.*

*Artıklar üzerine tek bir tahminsel model formunda olan modeller serisi kurulur.*

*Adaptive Boosting(Adaboost) bir optimizasyon algoritmasıdır: Zayıf sınıflandırıcıların bir araya gelerek güçlü bir sınıflandırıcı oluşturması fikrini hayata geçiren algoritmadır.*

*Yukarıda bir sınıflandırma işleminin görsel temsili verilmiş. D1 kutucuğu mavi sınıf ve kırmızı sınıf şeklinde iki sınıftan oluşuyor. Bazı değerler yanlış sınıflarda bulunuyor. Burada + değerler mavi sınıf değerlerini; - değerler kırmızı sınıf değerlerini temsil ediyor. Her bir iterasyonda farklı değerlere ağırlık verilerek tekrar bir sınıflandırma işlemi yapılıyor ve en sonda sınıflar doğru bir şekilde parçalara ayrılarak yanlış sınıflandırılan değerlerin doğru sınıflara geçmesi sağlanıyor.*

*GBM'de ağaçların birbirine bağımlılığı söz konusudur, yukarıda yaptığımız işlem kümülatif işlemdir.*

- *Gradient Boosting, tek bir tahminsel model formunda olan modeller serisi oluşturur.*
- *Seri içerisinde bir model serideki bir önceki modelin tahmin artıklarının / hatalarının(residuals) üzerine kurularak(fit) oluşturulur.*
- *GBM diferansiyellenebilen herhangi bir kayıp fonksiyonunu optimize edebilen Gradient Descent algoritmasını kullanmaktadır.*
- *GB bir çok temel öğrenici tipini (base learner type) kullanabilir. (Trees, linear terms, splines...)*
- *Cost Fonksiyonları ve Link Fonksiyonları modifiye edilebilir.*
- *Boosting + Gradient Descent olarak da bilinir.*

*GBM'yi; ağaca dayalı yöntemleri, lineer yöntemleri ölçeklendirebilir hale getirebilen teorik bir framework olarak görebiliriz.*

*Bu bölümde bundan sonra ele alacağımız algoritmaların temelinde GBM vardır. Hatta bu algoritmalar GBM'in performans iyileştirmelerine tabi tutulmuş daha iyi halleri de diyebiliriz.*
